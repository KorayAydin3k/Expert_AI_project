### 2. Helper Functions (Reusable, Leakage-Safe)


```python
# Helper functions (EDA only)

def get_numeric_features(df, exclude_cols, categorical_like_cols):
    numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
    return [c for c in numeric_cols if c not in exclude_cols + categorical_like_cols]


def top_skewed_features(df, numeric_features, k=6):
    skew_vals = df[numeric_features].skew(numeric_only=True).abs().sort_values(ascending=False)
    return skew_vals.head(k).index.tolist()


def clip_bounds(series, lower=0.01, upper=0.99):
    lo = series.quantile(lower)
    hi = series.quantile(upper)
    return lo, hi


def top_categories_with_other(df, col, top_n=10):
    vc = df[col].value_counts(dropna=False)
    top = vc.head(top_n).index
    return df[col].where(df[col].isin(top), other='Other')
```

**What this shows:** standardized helpers for clean, repeatable EDA.  
**Why it matters for modeling:** ensures consistent feature selection and category handling without leakage.

- **Target** is defined as **next-year revenue change** to avoid same-year leakage.


```python
# --- Configuration ---
TARGET_MODE = 'log_diff'  # 'log_diff' or 'revenue_change_next'
TARGET_COL = 'target_log_diff' if TARGET_MODE == 'log_diff' else 'revenue_change_next'  # training target
USE_CAPPED_TARGET_TRAIN = True  # cap revenue_change_next for training when TARGET_MODE != 'log_diff'

TARGET_SHIFT = 1  # target year = fiscal_year + 1 when using revenue_change_next_year
EPS = 1e-9
EPS_DENOM = 1.0  # min denominator for % change conversion (avoid scaling errors)
NEG_CLIP_TOL = 1e-6  # treat tiny negatives as rounding
REV_CHANGE_CAP_LO = -80  # cap extreme YoY revenue change (lower, percent points)
REV_CHANGE_CAP_HI = 200   # cap extreme YoY revenue change (upper, percent points)
NEAR_ZERO_THR = 5  # threshold for near-zero targets (percent points)
WINSOR_LO = 0.05  # winsorize target lower quantile
WINSOR_HI = 0.95  # winsorize target upper quantile
WEIGHT_EPS = 20.0  # downweight extremes in training
WEIGHT_CLIP_LO = 0.05
WEIGHT_CLIP_HI = 5.0
TARGET_TRANSFORM = 'signed_log'  # stabilize heavy-tail target
JUMP_ABS_THR = 300  # abs jump threshold for lagged growth
JUMP_POS_THR = 300  # positive jump threshold
JUMP_NEG_THR = -60  # negative jump threshold
COVID_DROP_THR = -20  # covid shock drop threshold (percent points)
COVID_GROW_THR = 20   # covid shock growth threshold (percent points)
BASE_SMALL_Q = 0.20  # quantile for small-base flag (train-only)
REV_JUMP_THR = 200  # flag company-year if revenue_change_clean exceeds this
FIN_GROWTH_PCT_CAP_LO = -2.0  # cap other financial YoY % for rev-jump companies
FIN_GROWTH_PCT_CAP_HI = 2.0

# Regime split / evaluation thresholds
EXTREME_THR = 75  # percent points; set to 300 if you want a wider 'normal' band
CLIP_EVAL_THR = 75  # metrics on |y| <= this
TRAIN_NORMAL_ONLY = False  # train only on |y| <= TRAIN_CLIP_THR (normal band)
TRAIN_CLIP_THR = CLIP_EVAL_THR
EVAL_CLIP_ONLY = False  # if True, eval metrics only on |y| <= CLIP_EVAL_THR

# Feature selection
SCALE_FEATURE_POLICY = 'raw'   # 'log', 'raw', or 'both'  (no log transforms - apply consistently or not at all)
FEATURE_POLICY = 'core'  # 'core', 'auto', 'all'
MISSINGNESS_MAX = 0.40  # for FEATURE_POLICY='auto'

RAW_ITEMS = [
    'total_fixed_assets', 'current_assets', 'total_assets',
    'shareholders_equity', 'total_debt', 'short_term_debt', 'long_term_debt',
    'production_value', 'production_costs', 'operating_income',
    'financial_income', 'financial_expenses', 'net_profit_loss'
]

RATIO_COLS = [
    'roe', 'roi', 'leverage', 'current_ratio', 'quick_ratio',
    'debt_to_assets', 'profit_margin'
]

NONNEG_COLS = [
    'total_fixed_assets', 'current_assets', 'total_assets',
    'total_debt', 'short_term_debt', 'long_term_debt',
    'production_value', 'production_costs',
    'financial_income', 'financial_expenses'
]

ID_COLS = ['company_id', 'fiscal_year']
LEAKAGE_COLS = ['bankruptcy_next_year', 'financial_health_class']

VALID_LEGAL_FORMS = {'SRL', 'SPA', 'SAPA', 'SAS', 'SNC'}

def get_covid_impact_score(ateco_code):
    """
    Returns 1 (low), 2 (medium), 3 (high) COVID impact score by ATECO main sector.
    Fold-safe: rule-based, no target info used.
    """
    try:
        code = str(ateco_code)
        sector_num = int(code[:2])
    except Exception:
        return 2

    # High impact: accommodation/food, arts/entertainment, transport/storage
    if sector_num in [55, 56, 90, 91, 92, 93, 49, 50, 51, 52, 53]:
        return 3

    # Medium impact: manufacturing, construction, wholesale/retail
    if (10 <= sector_num <= 33) or (41 <= sector_num <= 43) or (45 <= sector_num <= 47):
        return 2

    # Low impact: info/comm, health/social, agriculture, utilities/water
    if (58 <= sector_num <= 63) or (86 <= sector_num <= 88) or (1 <= sector_num <= 3) or (35 <= sector_num <= 39):
        return 1

    return 2



def add_covid_features(df):
    '''
    Add COVID-related features that are time-consistent for next-year targets.
    Uses only year t information (no future data).
    '''
    df = df.copy()

    if 'fiscal_year' in df.columns:
        df['is_covid_year'] = df['fiscal_year'].isin([2020, 2021]).astype(int)

    if 'ateco_sector' in df.columns:
        df['covid_impact_score'] = df['ateco_sector'].apply(get_covid_impact_score)
    else:
        df['covid_impact_score'] = np.nan

    # Shock/recovery signals use lagged revenue change from t-1 -> t (available at year t)
    if 'revchg_lag1' in df.columns and 'fiscal_year' in df.columns:
        is_2020 = df['fiscal_year'] == 2020
        df['is_big_drop_2020'] = (is_2020 & (df['revchg_lag1'] < COVID_DROP_THR)).astype(int)
        df['is_big_growth_2020'] = (is_2020 & (df['revchg_lag1'] > COVID_GROW_THR)).astype(int)

        shock_mag = (-df['revchg_lag1']).clip(lower=0)
        df['covid_shock_severity'] = np.where(is_2020, shock_mag, 0.0)

        # Interaction terms to let model learn rebound behavior
        df['covid_impact_x_lag1'] = np.where(is_2020, df['covid_impact_score'] * df['revchg_lag1'], 0.0)
        df['covid_recovery_potential'] = np.where(is_2020, df['covid_impact_score'] * shock_mag, 0.0)
    else:
        df['is_big_drop_2020'] = 0
        df['is_big_growth_2020'] = 0
        df['covid_shock_severity'] = 0.0
        df['covid_impact_x_lag1'] = 0.0
        df['covid_recovery_potential'] = 0.0

    return df

def pv_next_to_logdiff(pv_now, pv_next):
    """Compute log-diff target: log1p(pv_next) - log1p(pv_now).
    """
    return np.log1p(pv_next) - np.log1p(pv_now)

def logdiff_to_pv_next(pv_now, log_diff):
    """Invert log-diff to next-year production value.
    pred_next = expm1(log_diff + log1p(pv_now))
    """
    return np.expm1(log_diff + np.log1p(pv_now))

def std_str(s):
    return (
        s.astype('string')
         .str.strip()
         .str.upper()
         .replace({'': pd.NA, 'NAN': pd.NA, 'NONE': pd.NA})
    )

def standardize_legal_form(series):
    if series is None:
        return series, None
    s = std_str(series)
    repl = {
        'S.R.L.': 'SRL', 'S R L': 'SRL', 'SRL': 'SRL',
        'S.P.A.': 'SPA', 'S P A': 'SPA', 'SPA': 'SPA',
        'S.A.P.A.': 'SAPA', 'S A P A': 'SAPA', 'SAPA': 'SAPA',
        'S.A.S.': 'SAS', 'S A S': 'SAS', 'SAS': 'SAS',
        'S.N.C.': 'SNC', 'S N C': 'SNC', 'SNC': 'SNC'
    }
    s = s.replace(repl)
    is_other = ~s.isin(list(VALID_LEGAL_FORMS))
    s = s.where(~is_other, other='OTHER/UNKNOWN')
    return s, is_other.astype(int)

def safe_divide(a, b, eps=EPS):
    b_safe = b.where(~b.isna(), other=np.nan)
    return a / (b_safe.replace({0: np.nan}) + eps)

def resolve_duplicates(df, key_cols):
    dup_mask = df.duplicated(key_cols, keep=False)
    if not dup_mask.any():
        return df, 0
    tmp = df.copy()
    tmp['_non_nulls'] = tmp.notna().sum(axis=1)
    tmp = tmp.sort_values(key_cols + ['_non_nulls'], ascending=[True, True, False])
    deduped = tmp.drop_duplicates(key_cols, keep='first').drop(columns=['_non_nulls'])
    return deduped, int(dup_mask.sum())

def check_identity(lhs, rhs, tol=1e-3):
    diff = (lhs - rhs).abs()
    scale = np.maximum(1.0, lhs.abs())
    return diff <= tol * scale

def apply_accounting_fixes(df):
    df = df.copy()
    cols = df.columns

    # Fill totals from components when totals are missing
    if {'total_fixed_assets', 'current_assets', 'total_assets'} <= set(cols):
        mask = df['total_assets'].isna() & df['total_fixed_assets'].notna() & df['current_assets'].notna()
        df.loc[mask, 'total_assets'] = df.loc[mask, 'total_fixed_assets'] + df.loc[mask, 'current_assets']

    if {'short_term_debt', 'long_term_debt', 'total_debt'} <= set(cols):
        mask = df['total_debt'].isna() & df['short_term_debt'].notna() & df['long_term_debt'].notna()
        df.loc[mask, 'total_debt'] = df.loc[mask, 'short_term_debt'] + df.loc[mask, 'long_term_debt']

    # Identity checks
    a_ok = pd.Series(False, index=df.index)
    b_ok = pd.Series(False, index=df.index)
    c_ok = pd.Series(False, index=df.index)

    if {'total_assets', 'total_fixed_assets', 'current_assets'} <= set(cols):
        a_ok = check_identity(df['total_assets'], df['total_fixed_assets'] + df['current_assets'])
    if {'total_debt', 'short_term_debt', 'long_term_debt'} <= set(cols):
        b_ok = check_identity(df['total_debt'], df['short_term_debt'] + df['long_term_debt'])
    if {'total_assets', 'shareholders_equity', 'total_debt'} <= set(cols):
        c_ok = check_identity(df['total_assets'], df['shareholders_equity'] + df['total_debt'])

    # If A & B pass but C fails, derive equity
    if {'total_assets', 'shareholders_equity', 'total_debt'} <= set(cols):
        mask = a_ok & b_ok & (~c_ok) & df['total_assets'].notna() & df['total_debt'].notna()
        df.loc[mask, 'shareholders_equity'] = df.loc[mask, 'total_assets'] - df.loc[mask, 'total_debt']

    # Recompute checks after fixes
    if {'total_assets', 'total_fixed_assets', 'current_assets'} <= set(cols):
        a_ok = check_identity(df['total_assets'], df['total_fixed_assets'] + df['current_assets'])
    if {'total_debt', 'short_term_debt', 'long_term_debt'} <= set(cols):
        b_ok = check_identity(df['total_debt'], df['short_term_debt'] + df['long_term_debt'])
    if {'total_assets', 'shareholders_equity', 'total_debt'} <= set(cols):
        c_ok = check_identity(df['total_assets'], df['shareholders_equity'] + df['total_debt'])

    df['identity_A_ok'] = a_ok.astype(int)
    df['identity_B_ok'] = b_ok.astype(int)
    df['identity_C_ok'] = c_ok.astype(int)
    df['accounting_identity_violation'] = (~a_ok | ~b_ok | ~c_ok).astype(int)
    return df

def apply_negative_rules(df, cols):
    df = df.copy()
    for c in cols:
        if c not in df.columns:
            continue
        neg_mask = df[c] < -NEG_CLIP_TOL
        small_neg = (df[c] >= -NEG_CLIP_TOL) & (df[c] < 0)
        df.loc[small_neg, c] = 0
        df.loc[neg_mask, c] = np.nan
        df[f'neg_{c}'] = neg_mask.astype(int)
    return df

def add_ratio_flags(df):
    df = df.copy()
    # Replace inf with NaN for ratio columns
    ratio_cols = [c for c in RATIO_COLS if c in df.columns]
    if ratio_cols:
        df[ratio_cols] = df[ratio_cols].replace([np.inf, -np.inf], np.nan)

    # Invalid ratio conditions (denominator <= 0)
    if 'roe' in df.columns and 'shareholders_equity' in df.columns:
        df['roe_invalid'] = (df['shareholders_equity'] <= 0).astype(int)
    if 'roi' in df.columns and 'total_assets' in df.columns:
        df['roi_invalid'] = (df['total_assets'] <= 0).astype(int)
    if 'profit_margin' in df.columns and 'production_value' in df.columns:
        df['profit_margin_invalid'] = (df['production_value'] <= 0).astype(int)
    if 'current_ratio' in df.columns and 'short_term_debt' in df.columns:
        df['current_ratio_invalid'] = (df['short_term_debt'] <= 0).astype(int)
    if 'quick_ratio' in df.columns and 'short_term_debt' in df.columns:
        df['quick_ratio_invalid'] = (df['short_term_debt'] <= 0).astype(int)
    if 'debt_to_assets' in df.columns and 'total_assets' in df.columns:
        df['debt_to_assets_invalid'] = (df['total_assets'] <= 0).astype(int)
    if 'leverage' in df.columns and 'shareholders_equity' in df.columns:
        df['leverage_invalid'] = (df['shareholders_equity'] <= 0).astype(int)
    return df

def add_missing_indicators(df, cols):
    df = df.copy()
    for c in cols:
        if c in df.columns:
            df[f'is_missing_{c}'] = df[c].isna().astype(int)
    return df
```

```python
# --- Panel Cleaning + Feature Engineering (fold-safe) ---
# Real-life / no-peeking: do splitting by year first, then run this on the
# years available in that fold (train years + the validation year).

def build_panel_features(df_in, verbose=True):
    TARGET_MODE = globals().get('TARGET_MODE', 'log_diff')
    df_clean = df_in.copy()

    # Types
    if 'company_id' in df_clean.columns:
        df_clean['company_id'] = df_clean['company_id'].astype(str)
    if 'fiscal_year' in df_clean.columns:
        df_clean['fiscal_year'] = pd.to_numeric(df_clean['fiscal_year'], errors='coerce').astype('Int64')
        df_clean['year_index'] = df_clean['fiscal_year'] - df_clean['fiscal_year'].min()

    # Standardize categoricals
    for col in ['province', 'region']:
        if col in df_clean.columns:
            df_clean[col] = std_str(df_clean[col])

    if 'ateco_sector' in df_clean.columns:
        # Keep as 2-digit string code
        df_clean['ateco_sector'] = df_clean['ateco_sector'].astype('Int64').astype('string')
        df_clean['ateco_sector'] = df_clean['ateco_sector'].str.zfill(2).fillna('UNKNOWN')

    if 'legal_form' in df_clean.columns:
        df_clean['legal_form'], df_clean['legal_form_is_other'] = standardize_legal_form(df_clean['legal_form'])

    if 'legal_form' in df_clean.columns:
        # Legal form flags (simple, leakage-safe)
        df_clean['unlimited_liability_form'] = df_clean['legal_form'].isin(['SNC', 'SAS']).astype(int)
        df_clean['corporate_form'] = df_clean['legal_form'].isin(['SPA', 'SAPA']).astype(int)

    # Resolve duplicate (company_id, fiscal_year)
    if {'company_id', 'fiscal_year'} <= set(df_clean.columns):
        df_clean, dup_rows = resolve_duplicates(df_clean, ['company_id', 'fiscal_year'])
        if verbose:
            print(f'Duplicate rows detected (pre-resolve): {dup_rows}')

    # Accounting identities
    df_clean = apply_accounting_fixes(df_clean)

    # Negative value rules
    df_clean = apply_negative_rules(df_clean, NONNEG_COLS)

    # Ratio validity + inf handling
    df_clean = add_ratio_flags(df_clean)

    # Missingness indicators (raw + ratio + years_in_business)
    missing_cols = [c for c in RAW_ITEMS + RATIO_COLS + ['years_in_business'] if c in df_clean.columns]
    df_clean = add_missing_indicators(df_clean, missing_cols)

    # Feature engineering (no target recompute; no shifting)
    if {'company_id', 'fiscal_year'} <= set(df_clean.columns):
        df_clean = df_clean.sort_values(['company_id', 'fiscal_year']).copy()

    # Target year metadata (for alignment checks)
    # Target year metadata (for alignment checks)
    if 'fiscal_year' in df_clean.columns:
        df_clean['target_year'] = df_clean['fiscal_year'] + TARGET_SHIFT

    # Compute revenue_change from production_value (percent points)
    if 'production_value' in df_clean.columns and 'company_id' in df_clean.columns:
        pv_lag1 = df_clean.groupby('company_id')['production_value'].shift(1)
        df_clean['revenue_change_clean'] = np.where(pv_lag1 > 0, (df_clean['production_value'] - pv_lag1) / pv_lag1 * 100.0, np.nan)
    else:
        df_clean['revenue_change_clean'] = np.nan

    # Revenue jump flag (current-year, leakage-safe)
    if 'revenue_change_clean' in df_clean.columns:
        df_clean['rev_jump_flag'] = (df_clean['revenue_change_clean'].abs() >= REV_JUMP_THR).fillna(False).astype(int)
    else:
        df_clean['rev_jump_flag'] = 0

    # Past log growth + reversion flags (fold-safe)
    if 'production_value' in df_clean.columns and 'company_id' in df_clean.columns:
        pv_lag1 = df_clean.groupby('company_id')['production_value'].shift(1)
        df_clean['past_log_growth'] = np.log1p(df_clean['production_value']) - np.log1p(pv_lag1)
        df_clean['missing_history_flag'] = df_clean['past_log_growth'].isna().astype(int)
        df_clean['past_log_growth'] = df_clean['past_log_growth'].fillna(0)
        df_clean['is_post_merger'] = (df_clean['past_log_growth'] > 0.69).astype(int)
        df_clean['is_post_crash'] = (df_clean['past_log_growth'] < -0.5).astype(int)
        df_clean['reversion_pressure'] = -1 * df_clean['past_log_growth']

    # Lagged revenue change for event features
    if 'company_id' in df_clean.columns:
        df_clean['revchg_lag1'] = df_clean.groupby('company_id')['revenue_change_clean'].shift(1)
        df_clean['revchg_lag2'] = df_clean.groupby('company_id')['revenue_change_clean'].shift(2)

    # COVID-related features (time-consistent for next-year target)
    df_clean = add_covid_features(df_clean)

    # Jump/event flags based on lagged growth
    if 'revchg_lag1' in df_clean.columns:
        df_clean['jump_flag'] = (df_clean['revchg_lag1'].abs() > JUMP_ABS_THR).astype(int)
        df_clean['big_positive_jump'] = (df_clean['revchg_lag1'] > JUMP_POS_THR).astype(int)
        df_clean['big_negative_jump'] = (df_clean['revchg_lag1'] < JUMP_NEG_THR).astype(int)

    # Base-size normalization (lagged)
    if 'production_value' in df_clean.columns and 'company_id' in df_clean.columns:
        pv_lag1 = df_clean.groupby('company_id')['production_value'].shift(1)
        df_clean['log_pv_lag1'] = np.log1p(pv_lag1.clip(lower=0))

    # Structural change raw score (lagged deltas)
    if 'company_id' in df_clean.columns:
        # cost ratio and lag
        if {'production_costs', 'production_value'} <= set(df_clean.columns):
            df_clean['cost_ratio'] = safe_divide(df_clean['production_costs'], df_clean['production_value'])
            df_clean['cost_ratio_lag1'] = df_clean.groupby('company_id')['cost_ratio'].shift(1)
        # deltas
        if 'total_assets' in df_clean.columns:
            ta_lag1 = df_clean.groupby('company_id')['total_assets'].shift(1)
            df_clean['delta_total_assets'] = df_clean['total_assets'] - ta_lag1
        if 'total_fixed_assets' in df_clean.columns:
            tfa_lag1 = df_clean.groupby('company_id')['total_fixed_assets'].shift(1)
            df_clean['delta_total_fixed_assets'] = df_clean['total_fixed_assets'] - tfa_lag1
        if 'total_debt' in df_clean.columns:
            td_lag1 = df_clean.groupby('company_id')['total_debt'].shift(1)
            df_clean['delta_total_debt'] = df_clean['total_debt'] - td_lag1
        if 'cost_ratio' in df_clean.columns and 'cost_ratio_lag1' in df_clean.columns:
            df_clean['delta_cost_ratio'] = df_clean['cost_ratio'] - df_clean['cost_ratio_lag1']

        # aggregate structural change score (raw)
        score_parts = []
        for col in ['delta_total_assets', 'delta_total_fixed_assets', 'delta_total_debt', 'delta_cost_ratio']:
            if col in df_clean.columns:
                score_parts.append(df_clean[col].abs())
        if score_parts:
            df_clean['structural_change_raw'] = sum(score_parts)
    # Flag/cap extreme revenue changes (diagnostic + optional modeling)
    df_clean['is_extreme_revenue_change'] = ((df_clean['revenue_change_clean'] < REV_CHANGE_CAP_LO) | (df_clean['revenue_change_clean'] > REV_CHANGE_CAP_HI)).astype(int)
    df_clean['revenue_change_capped'] = df_clean['revenue_change_clean'].clip(REV_CHANGE_CAP_LO, REV_CHANGE_CAP_HI)

    # Next-year targets built from production_value (t -> t+1)
    if 'company_id' in df_clean.columns and 'production_value' in df_clean.columns:
        pv_now = df_clean['production_value']
        pv_next = df_clean.groupby('company_id')['production_value'].shift(-1)
        df_clean['production_value_next'] = pv_next
        df_clean['log_pv'] = np.log1p(pv_now.clip(lower=0))
        df_clean['log_pv_next'] = np.log1p(pv_next.clip(lower=0))
        df_clean['target_log_diff'] = df_clean['log_pv_next'] - df_clean['log_pv']
        denom = np.maximum(pv_now, EPS_DENOM)
        df_clean['revenue_change_next'] = np.where(pv_next.notna(), (pv_next - pv_now) / denom * 100.0, np.nan)
        df_clean['revenue_change_next_capped'] = df_clean['revenue_change_next'].clip(REV_CHANGE_CAP_LO, REV_CHANGE_CAP_HI)

    else:
        df_clean['production_value_next'] = np.nan
        df_clean['log_pv'] = np.nan
        df_clean['log_pv_next'] = np.nan
        df_clean['target_log_diff'] = np.nan
        df_clean['revenue_change_next'] = np.nan
        df_clean['revenue_change_next_capped'] = np.nan

    # Set target column according to mode
    df_clean[TARGET_COL] = df_clean['target_log_diff'] if TARGET_MODE == 'log_diff' else df_clean['revenue_change_next']

    # Lag/YoY/Growth features
    lag_base = [c for c in RAW_ITEMS + RATIO_COLS if c in df_clean.columns]
    if 'company_id' in df_clean.columns:
        for c in lag_base:
            df_clean[f'{c}_lag1'] = df_clean.groupby('company_id')[c].shift(1)
            df_clean[f'{c}_lag2'] = df_clean.groupby('company_id')[c].shift(2)

    growth_cols = [c for c in RAW_ITEMS if c in df_clean.columns]
    if 'company_id' in df_clean.columns:
        for c in growth_cols:
            lag1 = df_clean.groupby('company_id')[c].shift(1)
            df_clean[f'{c}_yoy_change'] = df_clean[c] - lag1
            df_clean[f'{c}_yoy_pct'] = (df_clean[c] - lag1) / (lag1.abs() + EPS)

    # Cap YoY % for other financials when revenue jump is extreme (current-year flag)
    if 'rev_jump_flag' in df_clean.columns:
        any_capped = pd.Series(False, index=df_clean.index)
        mask_jump = df_clean['rev_jump_flag'] == 1
        for c in growth_cols:
            pct_col = f'{c}_yoy_pct'
            if pct_col in df_clean.columns:
                before = df_clean.loc[mask_jump, pct_col]
                capped = before.clip(FIN_GROWTH_PCT_CAP_LO, FIN_GROWTH_PCT_CAP_HI)
                any_capped.loc[mask_jump] |= (capped != before)
                df_clean.loc[mask_jump, pct_col] = capped
        df_clean['fin_growth_capped_flag'] = any_capped.astype(int)
    else:
        df_clean['fin_growth_capped_flag'] = 0

    # Rolling features (past only)
    roll_cols = [c for c in ['production_value', 'operating_income', 'total_assets', 'total_debt', 'net_profit_loss'] if c in df_clean.columns]
    if 'company_id' in df_clean.columns:
        for c in roll_cols:
            df_clean[f'{c}_roll3_mean'] = (
                df_clean.groupby('company_id')[c]
                       .rolling(window=3, min_periods=2)
                       .mean()
                       .reset_index(level=0, drop=True)
            )
            df_clean[f'{c}_roll3_std'] = (
                df_clean.groupby('company_id')[c]
                       .rolling(window=3, min_periods=2)
                       .std()
                       .reset_index(level=0, drop=True)
            )

    # Derived indicators (current-year only)
    if {'operating_income', 'production_value'} <= set(df_clean.columns):
        df_clean['operating_margin'] = safe_divide(df_clean['operating_income'], df_clean['production_value'])
    if {'net_profit_loss', 'production_value'} <= set(df_clean.columns):
        df_clean['net_margin'] = safe_divide(df_clean['net_profit_loss'], df_clean['production_value'])
    if {'production_value', 'total_assets'} <= set(df_clean.columns):
        df_clean['asset_turnover'] = safe_divide(df_clean['production_value'], df_clean['total_assets'])
    if {'shareholders_equity', 'total_assets'} <= set(df_clean.columns):
        df_clean['equity_ratio'] = safe_divide(df_clean['shareholders_equity'], df_clean['total_assets'])
    if {'total_debt', 'shareholders_equity'} <= set(df_clean.columns):
        df_clean['debt_to_equity'] = safe_divide(df_clean['total_debt'], df_clean['shareholders_equity'])
    if {'current_assets', 'short_term_debt'} <= set(df_clean.columns):
        df_clean['working_capital'] = df_clean['current_assets'] - df_clean['short_term_debt']
        if 'total_assets' in df_clean.columns:
            df_clean['working_capital_ratio'] = safe_divide(df_clean['working_capital'], df_clean['total_assets'])

    # Additional efficiency & health ratios (fold-safe)
    if {'production_value', 'total_assets'} <= set(df_clean.columns):
        df_clean['asset_turnover'] = safe_divide(df_clean['production_value'], df_clean['total_assets'])
    if {'production_value', 'total_fixed_assets'} <= set(df_clean.columns):
        df_clean['fixed_asset_turnover'] = safe_divide(df_clean['production_value'], df_clean['total_fixed_assets'])
    if {'production_value', 'production_costs'} <= set(df_clean.columns):
        df_clean['op_efficiency'] = safe_divide(df_clean['production_value'], df_clean['production_costs'])
    if {'operating_income', 'financial_expenses'} <= set(df_clean.columns):
        df_clean['interest_coverage'] = safe_divide(df_clean['operating_income'], df_clean['financial_expenses'])
    if {'net_profit_loss', 'total_assets'} <= set(df_clean.columns):
        df_clean['roa'] = safe_divide(df_clean['net_profit_loss'], df_clean['total_assets'])
    if {'short_term_debt', 'total_debt'} <= set(df_clean.columns):
        df_clean['short_term_debt_ratio'] = safe_divide(df_clean['short_term_debt'], df_clean['total_debt'])
    if {'operating_income', 'production_value'} <= set(df_clean.columns):
        df_clean['ebit_margin'] = safe_divide(df_clean['operating_income'], df_clean['production_value'])

    # === ALTMAN Z-SCORE (Modified Z' for private companies) ===
    # Z' = 0.717*X1 + 0.847*X2 + 3.107*X3 + 0.420*X4 + 0.998*X5
    # X1 = working_capital / total_assets       (liquidity)
    # X2 = net_profit_loss / total_assets       (retained earnings / profitability)
    # X3 = operating_income / total_assets      (= roi, operating efficiency)
    # X4 = shareholders_equity / total_debt     (solvency buffer)
    # X5 = production_value / total_assets      (= asset_turnover, revenue efficiency)
    if {'working_capital', 'total_assets', 'net_profit_loss', 'operating_income',
        'shareholders_equity', 'production_value'} <= set(df_clean.columns):
        ta = df_clean['total_assets'].clip(lower=EPS)
        td = df_clean['total_debt'].clip(lower=EPS) if 'total_debt' in df_clean.columns else (
            (df_clean.get('short_term_debt', 0) + df_clean.get('long_term_debt', 0)).clip(lower=EPS)
        )
        X1 = df_clean['working_capital'] / ta
        X2 = df_clean['net_profit_loss'] / ta
        X3 = df_clean['operating_income'] / ta
        X4 = df_clean['shareholders_equity'] / td
        X5 = df_clean['production_value'] / ta
        df_clean['altman_z'] = (0.717 * X1 + 0.847 * X2 + 3.107 * X3 + 0.420 * X4 + 0.998 * X5)
        df_clean['altman_z'] = df_clean['altman_z'].replace([np.inf, -np.inf], np.nan).clip(-20, 20)

    # === RATIO TREND FEATURES (change from x-1 to x) ===
    # Signals whether financial health is improving or deteriorating
    for col in ['roe', 'roi', 'profit_margin', 'debt_to_assets', 'current_ratio']:
        lag1_col = f'{col}_lag1'
        if col in df_clean.columns and lag1_col in df_clean.columns:
            df_clean[f'{col}_chg'] = df_clean[col] - df_clean[lag1_col]

    # Revenue acceleration: is growth speeding up or slowing down?
    if 'revchg_lag1' in df_clean.columns and 'revchg_lag2' in df_clean.columns:
        df_clean['revenue_trend'] = df_clean['revchg_lag1'] - df_clean['revchg_lag2']

    return df_clean
```

```python
# --- Step 12: Expanding Window CV (real-life / no-peeking) ---
# Folds requested (single validation year):
# (train 2018-2019 -> val 2020)

FOLDS = [
    {'train_years': [2018, 2019], 'val_year': 2020},
]

# No test_features used in validation
USE_TEST_FOR_VAL_2021 = False
if USE_TEST_FOR_VAL_2021:
    test_features = pd.read_csv('../data/processed/test_features.csv')
else:
    test_features = None


# Collapse rare categories using TRAIN only
def collapse_rare(train, df, col, min_count=50):
    if col not in train.columns or col not in df.columns:
        return df
    vc = train[col].value_counts(dropna=False)
    keep = vc[vc >= min_count].index
    df = df.copy()
    df[col] = df[col].astype('string')
    df[col] = df[col].where(df[col].isin(keep.astype('string')), other='OTHER')
    return df

# Winsorization bounds (train-only)
def fit_winsor_bounds(df, cols, lower=0.01, upper=0.99):
    bounds = {}
    for c in cols:
        s = df[c].dropna()
        if s.empty:
            continue
        bounds[c] = (s.quantile(lower), s.quantile(upper))
    return bounds

# Outlier flags (train-only bounds) - compute BEFORE clipping
def add_outlier_flags(df, bounds):
    df = df.copy()
    for c, (lo, hi) in bounds.items():
        df[f'is_outlier_{c}'] = ((df[c] < lo) | (df[c] > hi)).astype(int)
    return df

def apply_winsor(df, bounds):
    df = df.copy()
    for c, (lo, hi) in bounds.items():
        df[c] = df[c].clip(lo, hi)
    return df

# Log1p transforms for non-negative scale variables
LOG_COLS_CANDIDATES = [
    'total_assets', 'total_fixed_assets', 'current_assets', 'total_debt',
    'short_term_debt', 'long_term_debt', 'production_value', 'production_costs'
]

def add_log1p(df, cols):
    df = df.copy()
    for c in cols:
        df[f'log1p_{c}'] = np.log1p(df[c].clip(lower=0))
    return df



# Feature selection controls (train-only)
FEATURE_POLICY = globals().get('FEATURE_POLICY', 'core')
MISSINGNESS_MAX = globals().get('MISSINGNESS_MAX', 0.40)

CORE_FEATURES = [
    # === CURRENT YEAR INCOME STATEMENT (x) ===
    # total_debt dropped: equals short_term_debt + long_term_debt (perfect multicollinearity)
    # operating_income dropped: captured by roi ratio
    'production_value',
    'production_costs',
    'net_profit_loss',

    # === CURRENT YEAR BALANCE SHEET (x) ===
    'total_assets',
    'total_fixed_assets',
    'current_assets',
    'shareholders_equity',
    'short_term_debt',
    'long_term_debt',
    'years_in_business',

    # === CURRENT YEAR FINANCIAL RATIOS (x) ===
    # leverage dropped: redundant with debt_to_assets
    # quick_ratio dropped: nearly identical to current_ratio (no inventory data)
    # net_margin / operating_margin dropped: captured by profit_margin + roi
    'roe',
    'roi',
    'profit_margin',
    'debt_to_assets',
    'current_ratio',

    # === DERIVED RATIOS (computed in build_panel_features) ===
    'asset_turnover',
    'working_capital_ratio',
    'interest_coverage',
    'altman_z',

    # === GROWTH SIGNALS (raw %, no log transforms) ===
    # revenue_change dropped: same as revenue_change_clean from original CSV
    # past_log_growth dropped: same info as revenue_change_clean but in log form
    'revenue_change_clean',
    'revchg_lag1',
    'revchg_lag2',
    'revenue_trend',

    # === PRIOR YEAR FINANCIAL STATEMENTS (x-1) ===
    'roe_lag1',
    'roi_lag1',
    'profit_margin_lag1',
    'debt_to_assets_lag1',
    'current_ratio_lag1',
    'production_value_lag1',
    'net_profit_loss_lag1',
    'operating_income_lag1',
    'total_assets_lag1',

    # === RATIO TRENDS (change from x-1 to x) ===
    'roe_chg',
    'roi_chg',
    'profit_margin_chg',
    'debt_to_assets_chg',
    'current_ratio_chg',

    # === EVENT FLAGS ===
    'missing_history_flag',
    'rev_jump_flag',
    'is_covid_year',
    'covid_impact_score',

    # === TEMPORAL ===
    'year_index',
]


def select_feature_cols(df_train, cat_cols, drop_cols):
    feature_cols = [c for c in df_train.columns if c not in drop_cols]

    if FEATURE_POLICY == 'core':
        allowed = set([c for c in CORE_FEATURES if c in df_train.columns])
        # Always keep categoricals
        feature_cols = [c for c in feature_cols if c in allowed or c in cat_cols]
        # Also keep quality/flag features
        quality_cols = [
            c for c in df_train.columns
            if c.startswith('is_missing_')
            or c.startswith('neg_')
            or c.endswith('_invalid')
            or c.endswith('_violation')
            or c.startswith('identity_')
            or c == 'accounting_identity_violation'
        ]
        feature_cols = sorted(set(feature_cols) | set(quality_cols))

    # Prefer log1p versions for scale features
    RAW_LOG_PAIRS = [
        ('production_value', 'log1p_production_value'),
        ('production_costs', 'log1p_production_costs'),
        ('total_assets', 'log1p_total_assets'),
        ('current_assets', 'log1p_current_assets'),
        ('total_debt', 'log1p_total_debt'),
        ('short_term_debt', 'log1p_short_term_debt'),
        ('long_term_debt', 'log1p_long_term_debt'),
        ('total_fixed_assets', 'log1p_total_fixed_assets'),
    ]
    for raw, logc in RAW_LOG_PAIRS:
        if logc in feature_cols and raw in feature_cols:
            feature_cols.remove(raw)
        elif FEATURE_POLICY == 'auto':
        # Drop features with high missingness in TRAIN (train-only decision)
            num_candidates = [c for c in feature_cols if c not in cat_cols]
            miss = df_train[num_candidates].isna().mean()
            keep_num = miss[miss <= MISSINGNESS_MAX].index.tolist()
            keep_num = [c for c in keep_num if df_train[c].nunique(dropna=True) > 1]
            feature_cols = [c for c in feature_cols if c in keep_num or c in cat_cols]
    # Scale feature policy: drop raw or log duplicates if requested
    SCALE_FEATURE_POLICY = globals().get('SCALE_FEATURE_POLICY', 'raw')
    RAW_LOG_PAIRS_GLOBAL = [
        ('production_value', 'log1p_production_value'),
        ('production_costs', 'log1p_production_costs'),
        ('total_assets', 'log1p_total_assets'),
        ('current_assets', 'log1p_current_assets'),
        ('total_debt', 'log1p_total_debt'),
        ('short_term_debt', 'log1p_short_term_debt'),
        ('long_term_debt', 'log1p_long_term_debt'),
        ('total_fixed_assets', 'log1p_total_fixed_assets'),
    ]
    for raw, logc in RAW_LOG_PAIRS_GLOBAL:
        if raw in feature_cols and logc in feature_cols:
            if SCALE_FEATURE_POLICY == 'log':
                feature_cols.remove(raw)
            elif SCALE_FEATURE_POLICY == 'raw':
                feature_cols.remove(logc)

    return feature_cols

def build_matrices(df_train_in, df_val_in, df_test_in=None):
    df_train = df_train_in.copy()
    df_val = df_val_in.copy()
    df_test = df_test_in.copy() if df_test_in is not None else None

    # Identify feature columns
    drop_cols = ID_COLS + LEAKAGE_COLS + [TARGET_COL]

    # Explicitly drop revenue_change_* features from modeling (keep only for diagnostics)
    drop_cols = drop_cols + [c for c in ['revenue_change_capped', 'production_value_next', 'log_pv_next', 'target_log_diff', 'revenue_change_next', 'revenue_change_next_capped'] if c in df_train.columns]

    # Hard drop any future/target-looking columns (extra safety)
    future_like = [c for c in df_train.columns if ('target' in c or c.endswith('_next') or c.endswith('_next_year'))]
    drop_cols = drop_cols + future_like
    if 'province' in df_train.columns:
        drop_cols = drop_cols + ['province']

    # Categorical features (province is dropped to avoid high cardinality)
    cat_cols = [c for c in ['ateco_sector', 'region', 'legal_form'] if c in df_train.columns]

    # Collapse rare categories using TRAIN only
    for c in cat_cols:
        df_train = collapse_rare(df_train, df_train, c, min_count=50)
        df_val = collapse_rare(df_train, df_val, c, min_count=50)
        if df_test is not None:
            df_test = collapse_rare(df_train, df_test, c, min_count=50)

    # Numeric columns to impute/winsorize
    num_cols = df_train.select_dtypes(include=[np.number]).columns.tolist()
    num_cols = [c for c in num_cols if c not in drop_cols]
    num_cols = [c for c in num_cols if not c.startswith('is_missing_') and not c.endswith('_invalid') and not c.endswith('_violation') and not c.startswith('neg_')]

    # Group median imputation (train-only stats)
    group_cols = [c for c in ['ateco_sector', 'region'] if c in df_train.columns]
    global_median = df_train[num_cols].median(numeric_only=True)

    if group_cols:
        group_medians = df_train.groupby(group_cols)[num_cols].median(numeric_only=True)
        group_index_train = pd.MultiIndex.from_frame(df_train[group_cols])
        group_index_val = pd.MultiIndex.from_frame(df_val[group_cols]) if not df_val.empty else None
        group_index_test = pd.MultiIndex.from_frame(df_test[group_cols]) if df_test is not None else None
    else:
        group_medians = None
        group_index_train = group_index_val = group_index_test = None

    year_medians = df_train.groupby('fiscal_year')[num_cols].median(numeric_only=True) if 'fiscal_year' in df_train.columns else None

    def impute_df(df, group_medians, group_index, year_medians, global_median, num_cols):
        df = df.copy()
        if group_medians is not None and group_index is not None:
            for c in num_cols:
                df[c] = df[c].fillna(pd.Series(group_index.map(group_medians[c]), index=df.index))
        if year_medians is not None and 'fiscal_year' in df.columns:
            for c in num_cols:
                df[c] = df[c].fillna(df['fiscal_year'].map(year_medians[c]))
        df[num_cols] = df[num_cols].fillna(global_median)
        return df

    df_train = impute_df(df_train, group_medians, group_index_train, year_medians, global_median, num_cols)
    df_val = impute_df(df_val, group_medians, group_index_val, year_medians, global_median, num_cols)
    if df_test is not None:
        df_test = impute_df(df_test, group_medians, group_index_test, year_medians, global_median, num_cols)


    # Denominator floor for % change (fixed, do not rescale)
    eps_denom = EPS_DENOM

    # Final safety fill for any remaining missing values in numeric features
    for _df in [df_train, df_val] + ([df_test] if df_test is not None else []):
        if _df is None or _df.empty:
            continue
        _df[num_cols] = _df[num_cols].fillna(global_median)

    # Winsorize (train-only bounds)
    winsor_bounds = fit_winsor_bounds(df_train, num_cols, lower=0.01, upper=0.99)
    df_train = add_outlier_flags(df_train, winsor_bounds)
    df_val = add_outlier_flags(df_val, winsor_bounds)
    if df_test is not None:
        df_test = add_outlier_flags(df_test, winsor_bounds)

    df_train = apply_winsor(df_train, winsor_bounds)
    df_val = apply_winsor(df_val, winsor_bounds)
    if df_test is not None:
        df_test = apply_winsor(df_test, winsor_bounds)

    # Log1p transforms
    log_cols = [c for c in LOG_COLS_CANDIDATES if c in df_train.columns]
    df_train = add_log1p(df_train, log_cols)
    df_val = add_log1p(df_val, log_cols)
    if df_test is not None:
        df_test = add_log1p(df_test, log_cols)

    # One-hot encode categoricals (ensure no missing)
    for _df in [df_train, df_val] + ([df_test] if df_test is not None else []):
        if _df is None or _df.empty:
            continue
        for c in cat_cols:
            if c in _df.columns:
                _df[c] = _df[c].astype('string').fillna('OTHER')

    feature_cols = select_feature_cols(df_train, cat_cols, drop_cols)

    X_train = pd.get_dummies(df_train[feature_cols], columns=cat_cols, drop_first=False)
    X_val = pd.get_dummies(df_val[feature_cols], columns=cat_cols, drop_first=False)
    X_val = X_val.reindex(columns=X_train.columns, fill_value=0)

    X_test = None
    if df_test is not None:
        X_test = pd.get_dummies(df_test[feature_cols], columns=cat_cols, drop_first=False)
        X_test = X_test.reindex(columns=X_train.columns, fill_value=0)

    y_train = df_train[TARGET_COL].copy()
    # Optionally cap training target only (no leakage into val/test)
    if globals().get('USE_CAPPED_TARGET_TRAIN', False):
        if TARGET_MODE == 'log_diff' and {'production_value', 'revenue_change_next_capped'} <= set(df_train.columns):
            pv_now = df_train['production_value']
            rc_cap = df_train['revenue_change_next_capped']
            pv_next_cap = pv_now * (1.0 + rc_cap / 100.0)
            y_train = np.log1p(pv_next_cap.clip(lower=0)) - np.log1p(pv_now.clip(lower=0))
        elif TARGET_MODE != 'log_diff':
            y_train = y_train.clip(REV_CHANGE_CAP_LO, REV_CHANGE_CAP_HI)
    y_val = df_val[TARGET_COL].copy()
    y_test = df_test[TARGET_COL].copy() if df_test is not None else None

    # Optional robust scaling for linear models (fit on train only)
    from sklearn.preprocessing import RobustScaler

    scaler = RobustScaler()
    X_train_scaled = X_train.copy()
    X_val_scaled = X_val.copy()

    # Scale only original numeric columns (post one-hot columns are already numeric but not in num_cols)
    num_cols_model = [c for c in X_train.columns if c in num_cols]
    X_train_scaled[num_cols_model] = scaler.fit_transform(X_train[num_cols_model])

    if X_val is not None and len(X_val) > 0 and num_cols_model:
        X_val_scaled[num_cols_model] = scaler.transform(X_val[num_cols_model])
    else:
        X_val_scaled = X_val.copy()

    if X_test is not None:
        X_test_scaled = X_test.copy()
        if len(X_test) > 0 and num_cols_model:
            X_test_scaled[num_cols_model] = scaler.transform(X_test[num_cols_model])
    else:
        X_test_scaled = None

    pv_now_train = df_train['production_value'].copy() if 'production_value' in df_train.columns else None
    pv_now_val = df_val['production_value'].copy() if 'production_value' in df_val.columns else None
    pv_now_test = df_test['production_value'].copy() if (df_test is not None and 'production_value' in df_test.columns) else None

    return {
        'df_train': df_train,
        'df_val': df_val,
        'df_test': df_test,
        'X_train': X_train,
        'X_val': X_val,
        'X_test': X_test,
        'y_train': y_train,
        'y_val': y_val,
        'y_test': y_test,
        'X_train_scaled': X_train_scaled,
        'X_val_scaled': X_val_scaled,
        'X_test_scaled': X_test_scaled,
        'pv_now_train': pv_now_train,
        'pv_now_val': pv_now_val,
        'pv_now_test': pv_now_test,
        'eps_denom': eps_denom,
        'num_cols_model': num_cols_model,
        'feature_cols': feature_cols,
        'cat_cols': cat_cols,
    }


fold_artifacts = []
for fold in FOLDS:
    train_years = fold['train_years']
    val_year = fold['val_year']

    # Include val_year+1 for target computation (t+1), but do not train on it
    label_years = sorted(set(train_years + [val_year]))
    target_years = sorted(set(label_years + [val_year + TARGET_SHIFT]))
    full_year = pd.to_numeric(full_df['fiscal_year'], errors='coerce')
    df_avail = full_df[full_year.isin(target_years)].copy()

    # If target year exceeds train_data range, append test_features rows for evaluation only
    if max(target_years) > full_year.max() and test_features is not None:
        test_year = pd.to_numeric(test_features['fiscal_year'], errors='coerce')
        df_eval = test_features[test_year.isin(target_years)].copy()
        df_avail = pd.concat([df_avail, df_eval], ignore_index=True, sort=False)

    # Deterministic cleaning + past-only features using only available years
    df_avail = build_panel_features(df_avail, verbose=False)

    # Keep only rows with target present AND in label years
    df_model_fold = df_avail[df_avail[TARGET_COL].notna() & df_avail['fiscal_year'].isin(label_years)].copy()

    df_train_fold = df_model_fold[df_model_fold['fiscal_year'].isin(train_years)].copy()
    df_val_fold = df_model_fold[df_model_fold['fiscal_year'] == val_year].copy()

    # Sector-year median growth (train-only) and growth_excess using lag1
    if {'ateco_sector', 'fiscal_year', 'revenue_change_clean'} <= set(df_train_fold.columns):
        sector_median = (
            df_train_fold
            .groupby(['ateco_sector', 'fiscal_year'])['revenue_change_clean']
            .median()
        )

        def add_sector_median_lag1(df):
            df = df.copy()
            year_lag1 = df['fiscal_year'] - 1
            idx = list(zip(df['ateco_sector'], year_lag1))
            df['sector_median_growth_lag1'] = [sector_median.get(k, np.nan) for k in idx]
            if 'revchg_lag1' in df.columns:
                df['growth_excess'] = df['revchg_lag1'] - df['sector_median_growth_lag1']
            return df

        df_train_fold = add_sector_median_lag1(df_train_fold)
        df_val_fold = add_sector_median_lag1(df_val_fold)


    # Sector shock prior (train-only): typical lagged shock by sector

    # Company prior (train-only): median lagged growth
    if {'company_id', 'revenue_change_clean'} <= set(df_train_fold.columns):
        company_prior = df_train_fold.groupby('company_id')['revenue_change_clean'].median()
        df_train_fold['company_prior'] = df_train_fold['company_id'].map(company_prior)
        df_val_fold['company_prior'] = df_val_fold['company_id'].map(company_prior)
        # fallback to sector median if missing
        if 'sector_median_growth_lag1' in df_train_fold.columns:
            df_train_fold['company_prior'] = df_train_fold['company_prior'].fillna(df_train_fold['sector_median_growth_lag1'])
            df_val_fold['company_prior'] = df_val_fold['company_prior'].fillna(df_val_fold['sector_median_growth_lag1'])
    if {'ateco_sector', 'revchg_lag1'} <= set(df_train_fold.columns):
        sector_shock_prior = df_train_fold.groupby('ateco_sector')['revchg_lag1'].median()
        df_train_fold['sector_shock_prior'] = df_train_fold['ateco_sector'].map(sector_shock_prior)
        df_val_fold['sector_shock_prior'] = df_val_fold['ateco_sector'].map(sector_shock_prior)
        df_train_fold['company_vs_sector_shock'] = df_train_fold['revchg_lag1'] - df_train_fold['sector_shock_prior']
        df_val_fold['company_vs_sector_shock'] = df_val_fold['revchg_lag1'] - df_val_fold['sector_shock_prior']

    # Base-small flag (train-only threshold on log_pv_lag1)
    if 'log_pv_lag1' in df_train_fold.columns:
        thr = df_train_fold['log_pv_lag1'].quantile(BASE_SMALL_Q)
        df_train_fold['base_small_flag'] = (df_train_fold['log_pv_lag1'] < thr).astype(int)
        df_val_fold['base_small_flag'] = (df_val_fold['log_pv_lag1'] < thr).astype(int)

    # Structural change z-score (train-only)
    if 'structural_change_raw' in df_train_fold.columns:
        mu = df_train_fold['structural_change_raw'].mean()
        sd = df_train_fold['structural_change_raw'].std()
        if sd and sd > 0:
            df_train_fold['structural_change_z'] = (df_train_fold['structural_change_raw'] - mu) / sd
            df_val_fold['structural_change_z'] = (df_val_fold['structural_change_raw'] - mu) / sd
        else:
            df_train_fold['structural_change_z'] = 0.0
            df_val_fold['structural_change_z'] = 0.0
    art = build_matrices(df_train_fold, df_val_fold, df_test_in=None)
    art['fold'] = fold
    fold_artifacts.append(art)

    print('Fold -> train', train_years, 'val', val_year)
    print('Shapes ->', art['X_train'].shape, art['X_val'].shape)

# Use last fold as the active (closest to final train) for downstream modeling cells
ACTIVE_FOLD = len(fold_artifacts) - 1
active = fold_artifacts[ACTIVE_FOLD]

# Export to globals for downstream cells
for k, v in active.items():
    globals()[k] = v

# Keep compatibility with older variable names
X_test = active.get('X_test')
y_test = active.get('y_test')
df_model = pd.concat([active['df_train'], active['df_val']], axis=0, ignore_index=True)

target_col = TARGET_COL
print('Active fold:', active['fold'])
print('Final matrices (active fold):')
val_target_year = active['fold']['val_year'] + TARGET_SHIFT
print('Validation target year:', val_target_year)
print('X_train', X_train.shape, 'y_train', y_train.shape)
print('X_val', X_val.shape, 'y_val', y_val.shape)
```

### Baseline Model

```python
TARGET_MODE = globals().get('TARGET_MODE', 'log_diff')
import numpy as np
# Baseline Model (Ridge + Signed-Log Target) with Group-Median Imputer + One-Hot Encoder
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, RobustScaler
from sklearn.impute import SimpleImputer
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score


# Helper for log-diff -> revenue_change

def logdiff_to_revenue_change(pv_now, log_diff):
    pv_next = np.expm1(log_diff + np.log1p(pv_now))
    denom = np.maximum(pv_now, EPS_DENOM)
    return (pv_next - pv_now) / denom * 100.0

# Use raw splits (before one-hot) but after leakage-safe feature engineering
base_train = df_train.copy()
base_val = df_val.copy()

target_col = target_col  # already set in preprocessing
leakage_cols = ['company_id', 'bankruptcy_next_year', 'financial_health_class', target_col, 'revenue_change_clean', 'revenue_change_capped', 'revenue_change_next', 'target_log_diff', 'production_value_next', 'log_pv_next']

class GroupMedianImputer(BaseEstimator, TransformerMixin):
    def __init__(self, group_cols, num_cols):
        self.group_cols = group_cols
        self.num_cols = num_cols
        self.group_medians_ = None
        self.global_median_ = None

    def fit(self, X, y=None):
        X = X.copy()
        self.global_median_ = X[self.num_cols].median(numeric_only=True)
        if self.group_cols:
            self.group_medians_ = X.groupby(self.group_cols)[self.num_cols].median(numeric_only=True)
        else:
            self.group_medians_ = None
        return self

    def transform(self, X):
        X = X.copy()
        if self.group_cols and self.group_medians_ is not None:
            group_index = pd.MultiIndex.from_frame(X[self.group_cols])
            for c in self.num_cols:
                X[c] = X[c].fillna(pd.Series(group_index.map(self.group_medians_[c]), index=X.index))
        X[self.num_cols] = X[self.num_cols].fillna(self.global_median_)
        return X

# Define columns
cat_cols = [c for c in ['ateco_sector', 'legal_form', 'region', 'fiscal_year'] if c in base_train.columns]
cat_cols = [c for c in cat_cols if base_train[c].dtype == 'object' or str(base_train[c].dtype).startswith('string')]
# Ensure categoricals are object dtype (sklearn SimpleImputer compatibility)
for _df in [base_train, base_val]:
    for c in cat_cols:
        _df[c] = _df[c].astype('object')
num_cols = base_train.select_dtypes(include=[np.number]).columns.tolist()
num_cols = [c for c in num_cols if c not in leakage_cols]

# Group-based imputation uses sector/legal_form/region when available
group_cols = [c for c in ['ateco_sector', 'legal_form', 'region'] if c in base_train.columns]

# Build preprocessing pipeline
numeric_pipe = Pipeline(steps=[
    ('median_imputer', SimpleImputer(strategy='median')),
    ('scaler', RobustScaler())
])

# OneHotEncoder API differs by sklearn version
try:
    onehot = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
except TypeError:
    onehot = OneHotEncoder(handle_unknown='ignore', sparse=False)

categorical_pipe = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', onehot)
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_pipe, num_cols),
        ('cat', categorical_pipe, cat_cols)
    ],
    remainder='drop'
)

# Full pipeline: group-median impute -> one-hot/scaling -> linear regression
model = Pipeline(steps=[
    ('group_imputer', GroupMedianImputer(group_cols=group_cols, num_cols=num_cols)),
    ('preprocess', preprocessor),
    ('ridge', Ridge(alpha=10.0, random_state=42))
])

X_train_base = base_train.drop(columns=leakage_cols, errors='ignore')
X_val_base = base_val.drop(columns=leakage_cols, errors='ignore')

y_train_base = y_train.copy()
y_val_base = y_val.copy()

# Signed-log target transform for stability (only for percent target)
sign_log = lambda y: np.sign(y) * np.log1p(np.abs(y))
inv_sign_log = lambda y: np.sign(y) * (np.expm1(np.abs(y)))

if TARGET_MODE == 'log_diff':
    # Train on log-diff, evaluate on revenue_change_next
    y_train_t = y_train_base.to_numpy()
    model.fit(X_train_base, y_train_t)
    y_pred_t = model.predict(X_val_base)
    pv_now = df_val['production_value'].to_numpy()
    y_pred_eval = logdiff_to_revenue_change(pv_now, y_pred_t)
    y_true_eval = df_val['revenue_change_next'].to_numpy()
else:
    y_train_t = sign_log(y_train_base)
    model.fit(X_train_base, y_train_t)
    y_pred_t = model.predict(X_val_base)
    y_pred_eval = inv_sign_log(y_pred_t)
    y_true_eval = y_val_base.to_numpy()

print('Baseline Model Performance (Group Median + OHE):')
print(f'R^2 Score: {r2_score(y_true_eval, y_pred_eval):.4f}')
print(f'Mean Absolute Error: {mean_absolute_error(y_true_eval, y_pred_eval):.4f}')
print(f'RMSE: {mean_squared_error(y_true_eval, y_pred_eval, squared=False):.4f}')

```

```python
TARGET_MODE = globals().get('TARGET_MODE', 'log_diff')
# Target transform helpers (robust to heavy tails)

def y_transform(y):
    if TARGET_MODE == 'log_diff':
        return np.asarray(y)
    # signed log1p transform
    return np.sign(y) * np.log1p(np.abs(y))


def y_inverse(y_t):
    if TARGET_MODE == 'log_diff':
        return np.asarray(y_t)
    return np.sign(y_t) * (np.expm1(np.abs(y_t)))
```

```python
# Target winsorization (train-only) to reduce tail noise
def winsorize_target(y_train, y_val, lower=0.01, upper=0.99, hard_lo=None, hard_hi=None):
    lo_q, hi_q = y_train.quantile([lower, upper])
    lo = max(lo_q, hard_lo) if hard_lo is not None else lo_q
    hi = min(hi_q, hard_hi) if hard_hi is not None else hi_q
    return y_train.clip(lo, hi), y_val.clip(lo, hi), lo, hi
```

```python
# Advanced Models (Regression) with target transform and robust metrics
# NOTE: val/test contain NaNs; only NaN-capable models are used
if 'active' in globals() and isinstance(active, dict):
    _vy = active['fold']['val_year']
    print(f'Evaluating models: features year {_vy} -> target year {_vy + TARGET_SHIFT}')

# - Baselines: median, lag1, sector mean
# - Random Forest
# - Gradient Boosting
# - XGBoost (if available)

from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, HistGradientBoostingClassifier
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import PowerTransformer, QuantileTransformer

# Use scaled features if available
if 'X_train_scaled' in globals():
    X_tr = X_train_scaled
    X_va = X_val_scaled
else:
    X_tr = X_train
    X_va = X_val

# Winsorize target (train-only) to stabilize heavy tails
if 'winsorize_target' in globals():
    y_train_clip, y_val_clip, y_lo, y_hi = winsorize_target(y_train, y_val, lower=WINSOR_LO, upper=WINSOR_HI, hard_lo=REV_CHANGE_CAP_LO, hard_hi=REV_CHANGE_CAP_HI)
else:
    y_train_clip, y_val_clip = y_train.copy(), y_val.copy()
    y_lo, y_hi = None, None

# Target transform selection
EPS_DENOM_EFFECTIVE = EPS_DENOM
TARGET_TRANSFORM = globals().get('TARGET_TRANSFORM', 'signed_log')

def transform_target(y_train_clip, y_val_clip, method):
    if method == 'signed_log':
        return y_transform(y_train_clip), y_transform(y_val_clip), None
    if method == 'yeo_johnson':
        yj = PowerTransformer(method='yeo-johnson', standardize=True)
        return yj.fit_transform(y_train_clip.to_numpy().reshape(-1,1)).ravel(), yj.transform(y_val_clip.to_numpy().reshape(-1,1)).ravel(), yj
    return y_train_clip.to_numpy(), y_val_clip.to_numpy(), None

def logdiff_to_revenue_change(pv_now, log_diff):
    pv_next = logdiff_to_pv_next(pv_now, log_diff)
    denom = np.maximum(pv_now, EPS_DENOM_EFFECTIVE)
    return (pv_next - pv_now) / denom * 100.0

def inverse_target(preds_t, method, yj=None):
    if method == 'signed_log':
        return y_inverse(preds_t)
    if method == 'yeo_johnson' and yj is not None:
        return yj.inverse_transform(preds_t.reshape(-1,1)).ravel()
    return preds_t

# Transform target for training
# Quantile transform only continuous numeric features (avoid one-hot/binary flags)
if isinstance(X_tr, pd.DataFrame):
    X_tr_q = X_tr.copy()
    X_va_q = X_va.copy()
    if 'num_cols_model' in globals():
        num_cols_q = [c for c in X_tr.columns if c in num_cols_model]
    else:
        num_cols_q = [c for c in X_tr.columns if X_tr[c].nunique(dropna=True) > 2]
    # always include log-scaled continuous columns
    extra_cols = [c for c in X_tr.columns if c.startswith('log1p_') or c.startswith('log_pv')]
    num_cols_q = sorted(set(num_cols_q) | set(extra_cols))
    if num_cols_q:
        qt = QuantileTransformer(n_quantiles=min(1000, len(X_tr)), output_distribution='normal', random_state=RANDOM_STATE)
        X_tr_q[num_cols_q] = qt.fit_transform(X_tr[num_cols_q])
        X_va_q[num_cols_q] = qt.transform(X_va[num_cols_q])
else:
    qt = QuantileTransformer(n_quantiles=min(1000, X_tr.shape[0]), output_distribution='normal', random_state=RANDOM_STATE)
    X_tr_q = qt.fit_transform(X_tr)
    X_va_q = qt.transform(X_va)


Y_tr_t, Y_va_t, yj = transform_target(y_train_clip, y_val_clip, TARGET_TRANSFORM)

# Evaluation targets in percent points when training on log-diff
if TARGET_MODE == 'log_diff' and 'revenue_change_next' in df_val.columns:
    y_val_eval = df_val['revenue_change_next'].to_numpy()
    y_train_eval = df_train['revenue_change_next'].to_numpy()
    pv_now_val = df_val['production_value'].to_numpy()
else:
    y_val_eval = y_val.to_numpy() if hasattr(y_val, 'to_numpy') else y_val
    y_train_eval = y_train.to_numpy() if hasattr(y_train, 'to_numpy') else y_train
    pv_now_val = None


# Metrics helper

def eps_mape(y_true, y_pred, eps=0.02):
    denom = np.maximum(np.abs(y_true), eps)
    return np.mean(np.abs(y_true - y_pred) / denom) * 100


def mape_thresholded(y_true, y_pred, thr=NEAR_ZERO_THR):
    y_true = np.asarray(y_true)
    y_pred = np.asarray(y_pred)
    mask = np.abs(y_true) >= thr
    if mask.sum() == 0:
        return np.nan, 0.0
    mape = np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask])) * 100
    coverage = mask.mean() * 100
    return mape, coverage

def mape_safe(y_true, y_pred):
    y_true = np.asarray(y_true)
    y_pred = np.asarray(y_pred)
    mask = y_true != 0
    if mask.sum() == 0:
        return np.nan
    return np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask])) * 100


def smape(y_true, y_pred):
    denom = (np.abs(y_true) + np.abs(y_pred)) / 2.0
    return np.mean(np.where(denom == 0, 0, np.abs(y_true - y_pred) / denom)) * 100


def directional_accuracy(y_true, y_pred):
    y_true = np.asarray(y_true)
    y_pred = np.asarray(y_pred)
    mask = np.isfinite(y_true) & np.isfinite(y_pred)
    if mask.sum() == 0:
        return np.nan
    return np.mean(np.sign(y_true[mask]) == np.sign(y_pred[mask])) * 100


# Clipped metrics on |y| <= CLIP_EVAL_THR (to track performance on the bulk)
def clipped_metrics(y_true, y_pred, thr=CLIP_EVAL_THR):
    y_true = np.asarray(y_true)
    y_pred = np.asarray(y_pred)
    mask = np.isfinite(y_true) & np.isfinite(y_pred) & (np.abs(y_true) <= thr)
    if mask.sum() == 0:
        return {'RMSE_clip': np.nan, 'MAE_clip': np.nan, 'MAPE_clip': np.nan, 'Clip_cov(%)': 0.0}
    rmse = mean_squared_error(y_true[mask], y_pred[mask], squared=False)
    mae = mean_absolute_error(y_true[mask], y_pred[mask])
    mape = mape_safe(y_true[mask], y_pred[mask])
    return {'RMSE_clip': rmse, 'MAE_clip': mae, 'MAPE_clip': mape, 'Clip_cov(%)': mask.mean() * 100}


# Robust evaluation helpers (raw vs clipped)
TAIL_ABS_THR = 300.0  # % points threshold for tail metrics

def eval_regression_robust(y_true, y_pred, clip_lo=-300.0, clip_hi=300.0):
    y_true = np.asarray(y_true)
    y_pred = np.asarray(y_pred)
    mask = np.isfinite(y_true) & np.isfinite(y_pred)
    if mask.sum() == 0:
        return {'RMSE': np.nan, 'MAE': np.nan, 'SMAPE': np.nan, 'MAPE': np.nan, 'R2': np.nan, 'DirectionalAcc(%)': np.nan, 'RMSE_clipped': np.nan, 'MAE_clipped': np.nan, 'SMAPE_clipped': np.nan, 'MAPE_clipped': np.nan, 'Tail_RMSE': np.nan, 'Tail_MAE': np.nan}
    y_true = y_true[mask]; y_pred = y_pred[mask]
    raw = {'RMSE': mean_squared_error(y_true, y_pred, squared=False), 'MAE': mean_absolute_error(y_true, y_pred), 'SMAPE': smape(y_true, y_pred), 'MAPE': mape_safe(y_true, y_pred), 'R2': r2_score(y_true, y_pred), 'DirectionalAcc(%)': directional_accuracy(y_true, y_pred)}
    y_true_c = np.clip(y_true, clip_lo, clip_hi); y_pred_c = np.clip(y_pred, clip_lo, clip_hi)
    clipped = {'RMSE_clipped': mean_squared_error(y_true_c, y_pred_c, squared=False), 'MAE_clipped': mean_absolute_error(y_true_c, y_pred_c), 'SMAPE_clipped': smape(y_true_c, y_pred_c), 'MAPE_clipped': mape_safe(y_true_c, y_pred_c)}
    tail_mask = np.abs(y_true) > TAIL_ABS_THR
    tail = {'Tail_RMSE': mean_squared_error(y_true[tail_mask], y_pred[tail_mask], squared=False), 'Tail_MAE': mean_absolute_error(y_true[tail_mask], y_pred[tail_mask])} if tail_mask.sum() > 0 else {'Tail_RMSE': np.nan, 'Tail_MAE': np.nan}
    return {**raw, **clipped, **tail}


# Train-time sample weights to downweight extremes
sample_weights = None
try:
    if 'y_train_clip' in locals():
        sample_weights = 1.0 / (np.abs(y_train_clip) + WEIGHT_EPS)
        sample_weights = np.clip(sample_weights, WEIGHT_CLIP_LO, WEIGHT_CLIP_HI)
except Exception:
    sample_weights = None

def eval_regression(name, model, X_tr, y_tr_t, X_va, y_va_raw):
    try:
        model.fit(X_tr, y_tr_t, sample_weight=sample_weights)
    except TypeError:
        model.fit(X_tr, y_tr_t)
    preds_t = model.predict(X_va)
    preds_t = np.asarray(preds_t)
    if not np.isfinite(preds_t).all():
        med = np.nanmedian(preds_t)
        preds_t = np.nan_to_num(preds_t, nan=med, posinf=med, neginf=med)
    preds = inverse_target(preds_t, TARGET_TRANSFORM, yj)
    if TARGET_MODE == 'log_diff' and pv_now_val is not None:
        preds = logdiff_to_revenue_change(pv_now_val, preds)
        y_eval = y_val_eval
    else:
        y_eval = y_va_raw
    rmse = mean_squared_error(y_eval, preds, squared=False)
    mae = mean_absolute_error(y_eval, preds)
    mape = mape_safe(y_eval, preds)
    mape_thr, mape_cov = mape_thresholded(y_eval, preds, thr=NEAR_ZERO_THR)
    r2 = r2_score(y_eval, preds)
    clip = clipped_metrics(y_eval, preds, thr=CLIP_EVAL_THR)
    return {
        'Model': name,
        'RMSE': rmse,
        'MAE': mae,
        'MAPE': mape,
        'MAPE_thr': mape_thr,
        'MAPE_cov(%)': mape_cov,
        'SMAPE': smape(y_eval, preds),
        'R2': r2,
        'DirectionalAcc(%)': directional_accuracy(y_eval, preds),
        **clip
    }


results = []

# Baseline 1: predict train median (percent scale)
baseline_pred = np.full_like(y_val_eval, np.nanmedian(y_train_eval), dtype=float)
clip = clipped_metrics(y_val_eval, baseline_pred, thr=CLIP_EVAL_THR)
results.append({
    'Model': 'Baseline (median)',
    'RMSE': mean_squared_error(y_val_eval, baseline_pred, squared=False),
    'MAE': mean_absolute_error(y_val_eval, baseline_pred),
    'MAPE': mape_safe(y_val_eval, baseline_pred),
    'SMAPE': smape(y_val_eval, baseline_pred),
    'R2': r2_score(y_val_eval, baseline_pred),
    'DirectionalAcc(%)': directional_accuracy(y_val_eval, baseline_pred),
    **clip
})

# Baseline 2: naive persistence (use current year's change as forecast)
if 'revenue_change_clean' in df_val.columns:
    pred = df_val['revenue_change_clean'].fillna(np.nanmedian(y_train_eval))
    clip = clipped_metrics(y_val_eval, pred, thr=CLIP_EVAL_THR)
    results.append({
        'Model': 'Baseline (persistence)',
        'RMSE': mean_squared_error(y_val_eval, pred, squared=False),
        'MAE': mean_absolute_error(y_val_eval, pred),
        'MAPE': mape_safe(y_val_eval, pred),
        'SMAPE': smape(y_val_eval, pred),
        'R2': r2_score(y_val_eval, pred),
        'DirectionalAcc(%)': directional_accuracy(y_val_eval, pred),
        **clip
    })

# Baseline 3: sector mean of target (train only)
if 'ateco_sector' in df_train.columns:
    if TARGET_MODE == 'log_diff' and 'revenue_change_next' in df_train.columns:
        sector_mean = df_train.groupby('ateco_sector')['revenue_change_next'].mean()
    else:
        sector_mean = df_train.groupby('ateco_sector')[target_col].mean()
    pred = df_val['ateco_sector'].map(sector_mean).fillna(np.nanmedian(y_train_eval))
    clip = clipped_metrics(y_val_eval, pred, thr=CLIP_EVAL_THR)
    results.append({
        'Model': 'Baseline (sector mean)',
        'RMSE': mean_squared_error(y_val_eval, pred, squared=False),
        'MAE': mean_absolute_error(y_val_eval, pred),
        'MAPE': mape_safe(y_val_eval, pred),
        'SMAPE': smape(y_val_eval, pred),
        'R2': r2_score(y_val_eval, pred),
        'DirectionalAcc(%)': directional_accuracy(y_val_eval, pred),
        **clip
    })


from sklearn.ensemble import HistGradientBoostingRegressor

# HistGradientBoosting (NaN-capable)
hgb_model = HistGradientBoostingRegressor(
    max_depth=6,
    learning_rate=0.05,
    max_iter=300,
    loss='absolute_error',
    random_state=RANDOM_STATE
)
results.append(eval_regression('HistGradientBoosting', hgb_model, X_tr_q, Y_tr_t, X_va_q, y_val))

# XGBoost (if available)
try:
    from xgboost import XGBRegressor
    xgb_model = XGBRegressor(
        n_estimators=500,
        learning_rate=0.05,
        max_depth=4,
        subsample=0.8,
        colsample_bytree=0.8,
        objective='reg:absoluteerror',
        random_state=RANDOM_STATE,
        n_jobs=-1
    )
    results.append(eval_regression('XGBoost', xgb_model, X_tr_q, Y_tr_t, X_va_q, y_val))
except Exception as e:
    print(f'XGBoost not available or failed to import: {e}')

# RandomForest
rf_model = RandomForestRegressor(
    n_estimators=400,
    max_depth=None,
    min_samples_split=5,
    min_samples_leaf=2,
    max_features='sqrt',
    n_jobs=-1,
    random_state=RANDOM_STATE
)
results.append(eval_regression('RandomForest', rf_model, X_tr_q, Y_tr_t, X_va_q, y_val))




# --- Mixture of Experts (Soft Routing) ---
THRESH_LOW = -15.0  # percent points
THRESH_HIGH = 15.0  # percent points

def assign_bucket_val(y):
    if y < THRESH_LOW:
        return 0  # Bear
    if y > THRESH_HIGH:
        return 2  # Bull
    return 1  # Frog

# Prepare features (fill NaNs using train median)
X_tr_moe = X_train.copy()
X_va_moe = X_val.copy()
train_median = X_tr_moe.median(numeric_only=True)
X_tr_moe = X_tr_moe.fillna(train_median).fillna(0)
X_va_moe = X_va_moe.fillna(train_median).fillna(0)

# Align training target to feature index for MoE
if hasattr(y_train, 'loc'):
    y_train_moe = y_train.loc[X_tr_moe.index]
else:
    y_train_moe = pd.Series(y_train, index=X_tr_moe.index)

if TARGET_MODE == 'log_diff':
    # bucket by percent-scale target, but train on log-diff
    y_bucket_tr = pd.Series(y_train_eval, index=X_tr_moe.index).apply(assign_bucket_val)
else:
    y_bucket_tr = y_train_moe.apply(assign_bucket_val)

# Train router (3-class)
moe_ready = y_bucket_tr.nunique() >= 2
if moe_ready:
    router = HistGradientBoostingClassifier(max_depth=4, learning_rate=0.05, max_iter=300, random_state=RANDOM_STATE)
    router.fit(X_tr_moe, y_bucket_tr)
    probs = router.predict_proba(X_va_moe)

    # Train experts on signed-log target
    experts = {}
    default_pred = float(y_train.median())
    for bucket_id in [0, 1, 2]:
        mask = (y_bucket_tr == bucket_id).to_numpy()
        if mask.sum() < 50:
            experts[bucket_id] = None
            continue
        reg = HistGradientBoostingRegressor(max_depth=6, learning_rate=0.05, max_iter=400, random_state=RANDOM_STATE, loss='absolute_error')
        reg.fit(X_tr_moe.iloc[mask], y_transform(y_train_moe.iloc[mask]))
        experts[bucket_id] = reg

    # Expert predictions for all val rows
    pred_0 = default_pred if experts[0] is None else y_inverse(experts[0].predict(X_va_moe))
    pred_1 = default_pred if experts[1] is None else y_inverse(experts[1].predict(X_va_moe))
    pred_2 = default_pred if experts[2] is None else y_inverse(experts[2].predict(X_va_moe))

    moe_pred = (probs[:, 0] * pred_0) + (probs[:, 1] * pred_1) + (probs[:, 2] * pred_2)

    if TARGET_MODE == 'log_diff' and pv_now_val is not None:
        moe_pred_eval = logdiff_to_revenue_change(pv_now_val, moe_pred)
        y_eval = y_val_eval
    else:
        moe_pred_eval = moe_pred
        y_eval = y_val

    results.append({
        'Model': 'Mixture-of-Experts (soft)',
        'RMSE': mean_squared_error(y_eval, moe_pred_eval, squared=False),
        'MAE': mean_absolute_error(y_eval, moe_pred_eval),
        'MAPE': mape_safe(y_eval, moe_pred_eval),
        'SMAPE': smape(y_eval, moe_pred_eval),
        'R2': r2_score(y_eval, moe_pred_eval),
        'DirectionalAcc(%)': directional_accuracy(y_eval, moe_pred_eval),
        'MAPE_thr': mape_thresholded(y_eval, moe_pred_eval, thr=NEAR_ZERO_THR)[0],
        'MAPE_cov(%)': mape_thresholded(y_eval, moe_pred_eval, thr=NEAR_ZERO_THR)[1]
    })
else:
    print('MoE skipped: not enough bucket diversity in training data.')

results_df = pd.DataFrame(results)
# Headline metrics: SMAPE + Directional Acc
results_df = results_df.sort_values(['SMAPE', 'DirectionalAcc(%)'])
print('Advanced models evaluated on validation set (original scale):')
display(results_df)


# LightGBM (if available)
try:
    from lightgbm import LGBMRegressor
    lgb_model = LGBMRegressor(
        n_estimators=1000,
        learning_rate=0.03,
        num_leaves=31,
        max_depth=-1,
        subsample=0.8,
        colsample_bytree=0.8,
        reg_lambda=1.0,
        objective='regression_l1',
        random_state=RANDOM_STATE
    )
    results.append(eval_regression('LightGBM', lgb_model, X_tr_q, Y_tr_t, X_va_q, y_val))
except Exception as e:
    print(f'LightGBM not available or failed: {e}')
```

```python

# --- Regime Split Model: Normal vs Extreme (|y| >= EXTREME_THR) ---
from sklearn.ensemble import HistGradientBoostingClassifier, HistGradientBoostingRegressor

# Use percent-scale targets for regime split
if TARGET_MODE == 'log_diff':
    y_tr_pct = y_train_eval
    y_va_pct = y_val_eval
else:
    y_tr_pct = y_train.to_numpy() if hasattr(y_train, 'to_numpy') else y_train
    y_va_pct = y_val.to_numpy() if hasattr(y_val, 'to_numpy') else y_val

# Features: reuse quantile-transformed matrices if available
X_tr_reg = X_tr_q if 'X_tr_q' in globals() else X_train
X_va_reg = X_va_q if 'X_va_q' in globals() else X_val

mask_ext_tr = np.abs(y_tr_pct) >= EXTREME_THR
mask_norm_tr = ~mask_ext_tr

print(f"Regime split threshold: {EXTREME_THR} | Train extreme share: {mask_ext_tr.mean()*100:.1f}%")

# Classifier: probability of extreme regime
if mask_ext_tr.mean() > 0.02 and mask_ext_tr.mean() < 0.98:
    clf = HistGradientBoostingClassifier(max_depth=4, learning_rate=0.05, max_iter=300, random_state=RANDOM_STATE)
    clf.fit(X_tr_reg, mask_ext_tr)
    p_ext = clf.predict_proba(X_va_reg)[:, 1]
else:
    # fallback to constant probability
    p_ext = np.full(len(y_va_pct), mask_ext_tr.mean())

# Regressors for each regime
if mask_norm_tr.sum() >= 200:
    reg_norm = HistGradientBoostingRegressor(max_depth=6, learning_rate=0.05, max_iter=500, random_state=RANDOM_STATE, loss='absolute_error')
    reg_norm.fit(X_tr_reg[mask_norm_tr], y_tr_pct[mask_norm_tr])
    pred_norm = reg_norm.predict(X_va_reg)
else:
    pred_norm = np.full(len(y_va_pct), np.nanmedian(y_tr_pct))

if mask_ext_tr.sum() >= 200:
    reg_ext = HistGradientBoostingRegressor(max_depth=4, learning_rate=0.05, max_iter=400, random_state=RANDOM_STATE, loss='absolute_error')
    reg_ext.fit(X_tr_reg[mask_ext_tr], y_tr_pct[mask_ext_tr])
    pred_ext = reg_ext.predict(X_va_reg)
else:
    pred_ext = np.full(len(y_va_pct), np.nanmedian(y_tr_pct))

# Soft blend
pred_regime = (1.0 - p_ext) * pred_norm + p_ext * pred_ext

# Evaluate
rmse = mean_squared_error(y_va_pct, pred_regime, squared=False)
mae = mean_absolute_error(y_va_pct, pred_regime)
mape = mape_safe(y_va_pct, pred_regime)
sm = smape(y_va_pct, pred_regime)
r2 = r2_score(y_va_pct, pred_regime)
da = directional_accuracy(y_va_pct, pred_regime)
clip = clipped_metrics(y_va_pct, pred_regime, thr=CLIP_EVAL_THR)

print('Regime Split (soft) results:')
print(f"RMSE: {rmse:.3f}  MAE: {mae:.3f}  MAPE: {mape:.3f}  SMAPE: {sm:.3f}  R2: {r2:.3f}  DirAcc: {da:.2f}%")
print(f"Clipped@{CLIP_EVAL_THR}: RMSE {clip['RMSE_clip']:.3f}  MAE {clip['MAE_clip']:.3f}  MAPE {clip['MAPE_clip']:.3f}  Coverage {clip['Clip_cov(%)']:.1f}%")
```

```python
# 3-stage bucket model (near-zero + direction + magnitude, leakage-safe)
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import HistGradientBoostingRegressor
from sklearn.metrics import accuracy_score
if 'active' in globals() and isinstance(active, dict):
    _vy = active['fold']['val_year']
    print(f'Bucket model: features year {_vy} -> target year {_vy + TARGET_SHIFT}')


thr = NEAR_ZERO_THR
# Use scaled features if available (more stable for linear models)
if 'X_train_scaled' in globals():
    X_tr_nz = X_train_scaled.copy()
    X_va_nz = X_val_scaled.copy()
else:
    X_tr_nz = X_train.copy()
    X_va_nz = X_val.copy()

train_median = X_tr_nz.median(numeric_only=True)
X_tr_nz = X_tr_nz.fillna(train_median).fillna(0)
X_va_nz = X_va_nz.fillna(train_median).fillna(0)

# Stage A: classify near-zero |y| < thr (train-only)
y_nz_tr = (y_train.abs() < thr).astype(int)
y_nz_va = (y_val.abs() < thr).astype(int)

if y_nz_tr.nunique() < 2:
    nz_pred = np.full_like(y_nz_va, y_nz_tr.iloc[0], dtype=int)
    nz_acc = (y_nz_va == nz_pred).mean()
else:
    nz_clf = LogisticRegression(max_iter=2000, class_weight='balanced')
    nz_clf.fit(X_tr_nz, y_nz_tr)
    nz_pred = nz_clf.predict(X_va_nz)
    nz_acc = accuracy_score(y_nz_va, nz_pred)

# Stage B: regress magnitude for non-near-zero (train-only)
y_train_mag = y_train_clip if 'y_train_clip' in globals() else y_train
mask_tr = y_train_mag.abs() >= thr
if mask_tr.sum() == 0:
    mag_pred = np.zeros_like(y_val, dtype=float)
else:
    mag_reg = HistGradientBoostingRegressor(max_depth=6, learning_rate=0.05, max_iter=400, random_state=RANDOM_STATE)
    mag_reg.fit(X_tr_nz[mask_tr], y_train_mag.abs()[mask_tr])
    mag_pred = mag_reg.predict(X_va_nz)

# Stage C: direction classifier (train-only)
if (y_train > 0).nunique() < 2:
    dir_pred = np.full_like(y_val, 1 if y_train.iloc[0] > 0 else -1, dtype=int)
else:
    dir_clf = LogisticRegression(max_iter=2000, class_weight='balanced')
    dir_clf.fit(X_tr_nz, (y_train > 0).astype(int))
    dir_pred = dir_clf.predict(X_va_nz)
    dir_pred = np.where(dir_pred == 1, 1, -1)

# Combine (leakage-safe)
y_pred_nz = np.where(nz_pred == 1, 0.0, dir_pred * mag_pred)

rmse = mean_squared_error(y_val, y_pred_nz, squared=False)
mae = mean_absolute_error(y_val, y_pred_nz)
mape = mape_safe(y_val, y_pred_nz) if 'mape_safe' in globals() else np.nan
mape_thr, mape_cov = mape_thresholded(y_val, y_pred_nz, thr=thr) if 'mape_thresholded' in globals() else (np.nan, 0.0)
sm = smape(y_val, y_pred_nz)
r2 = r2_score(y_val, y_pred_nz)
da = directional_accuracy(y_val, y_pred_nz)

print('Near-zero bucket model:')
print(f'Near-zero classifier accuracy: {nz_acc:.4f}')
print(f'RMSE: {rmse:.4f}  MAE: {mae:.4f}  MAPE: {mape:.4f}  MAPE_thr: {mape_thr:.4f} (cov {mape_cov:.1f}%)  SMAPE: {sm:.4f}  R2: {r2:.4f}  DirectionalAcc: {da:.4f}')
```

```python
TARGET_MODE = globals().get('TARGET_MODE', 'log_diff')
# Hyperparameter tuning (focused, leakage-safe) with target transform
# Tuning for RMSE + epsilon-MAPE (epsilon=0.02)
# NOTE: val/test have NaNs; use NaN-capable models only
if 'active' in globals() and isinstance(active, dict):
    _vy = active['fold']['val_year']
    print(f'Tuning models: features year {_vy} -> target year {_vy + TARGET_SHIFT}')

# Tune only on TRAIN data using time-aware CV; evaluate on VAL after tuning.

from sklearn.model_selection import RandomizedSearchCV, TimeSeriesSplit, GroupKFold, PredefinedSplit
from sklearn.metrics import make_scorer
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, HistGradientBoostingRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, median_absolute_error, r2_score
from sklearn.preprocessing import PowerTransformer, QuantileTransformer

# Choose scorer for tuning: 'mape' or 'mape_thr'
TUNING_SCORER = 'mape_thr'

# Use scaled features if available
if 'X_train_scaled' in globals():
    X_tr = X_train_scaled
    X_va = X_val_scaled
else:
    X_tr = X_train
    X_va = X_val

# Ensure no NaNs for estimators that don't accept missing values
# Use train medians for numeric columns; fill remaining with 0 as a final safety net
if isinstance(X_tr, pd.DataFrame):
    train_median = X_tr.median(numeric_only=True)
    X_tr_no_nan = X_tr.copy()
    X_va_no_nan = X_va.copy()
    X_tr_no_nan = X_tr_no_nan.fillna(train_median)
    X_va_no_nan = X_va_no_nan.fillna(train_median)
    X_tr_no_nan = X_tr_no_nan.fillna(0)
    X_va_no_nan = X_va_no_nan.fillna(0)
else:
    # numpy arrays
    X_tr_no_nan = np.nan_to_num(X_tr, nan=0.0)
    X_va_no_nan = np.nan_to_num(X_va, nan=0.0)

# Target transform selection
TARGET_TRANSFORM = globals().get('TARGET_TRANSFORM', 'signed_log')

def transform_target(y_train_clip, y_val_clip, method):
    if method == 'signed_log':
        return y_transform(y_train_clip), y_transform(y_val_clip), None
    if method == 'yeo_johnson':
        yj = PowerTransformer(method='yeo-johnson', standardize=True)
        return yj.fit_transform(y_train_clip.to_numpy().reshape(-1,1)).ravel(), yj.transform(y_val_clip.to_numpy().reshape(-1,1)).ravel(), yj
    return y_train_clip.to_numpy(), y_val_clip.to_numpy(), None

def logdiff_to_revenue_change(pv_now, log_diff):
    pv_next = np.expm1(log_diff + np.log1p(pv_now))
    denom = np.maximum(pv_now, EPS_DENOM)
    return (pv_next - pv_now) / denom * 100.0

def inverse_target(preds_t, method, yj=None):
    if method == 'signed_log':
        return y_inverse(preds_t)
    if method == 'yeo_johnson' and yj is not None:
        return yj.inverse_transform(preds_t.reshape(-1,1)).ravel()
    return preds_t

# Winsorize target (train-only) to stabilize heavy tails
# Quantile transform only continuous numeric features (avoid one-hot/binary flags)
if isinstance(X_tr_no_nan, pd.DataFrame):
    X_tr_no_nan = X_tr_no_nan.copy()
    X_va_no_nan = X_va_no_nan.copy()
    if 'num_cols_model' in globals():
        num_cols_q = [c for c in X_tr_no_nan.columns if c in num_cols_model]
    else:
        num_cols_q = [c for c in X_tr_no_nan.columns if X_tr_no_nan[c].nunique(dropna=True) > 2]
    extra_cols = [c for c in X_tr_no_nan.columns if c.startswith('log1p_') or c.startswith('log_pv')]
    num_cols_q = sorted(set(num_cols_q) | set(extra_cols))
    if num_cols_q:
        qt = QuantileTransformer(n_quantiles=min(1000, len(X_tr_no_nan)), output_distribution='normal', random_state=RANDOM_STATE)
        X_tr_no_nan[num_cols_q] = qt.fit_transform(X_tr_no_nan[num_cols_q])
        X_va_no_nan[num_cols_q] = qt.transform(X_va_no_nan[num_cols_q])
else:
    qt = QuantileTransformer(n_quantiles=min(1000, X_tr_no_nan.shape[0]), output_distribution='normal', random_state=RANDOM_STATE)
    X_tr_no_nan = qt.fit_transform(X_tr_no_nan)
    X_va_no_nan = qt.transform(X_va_no_nan)

if 'winsorize_target' in globals():
    y_train_clip, y_val_clip, y_lo, y_hi = winsorize_target(y_train, y_val, lower=WINSOR_LO, upper=WINSOR_HI, hard_lo=REV_CHANGE_CAP_LO, hard_hi=REV_CHANGE_CAP_HI)
else:
    y_train_clip, y_val_clip = y_train.copy(), y_val.copy()
    y_lo, y_hi = None, None

# For tuning, use raw (winsorized) target to avoid CV leakage from global transform

# Eval targets in percent points when training on log-diff
if TARGET_MODE == 'log_diff' and 'revenue_change_next' in df_val.columns:
    y_val_eval = df_val['revenue_change_next'].to_numpy()
    pv_now_val = df_val['production_value'].to_numpy()
else:
    y_val_eval = y_val.to_numpy() if hasattr(y_val, 'to_numpy') else y_val
    pv_now_val = None

Y_tr_t = y_train_clip
Y_va_t = y_val

# Optional: train only on normal band (by percent-scale target)
TRAIN_NORMAL_ONLY = globals().get('TRAIN_NORMAL_ONLY', False)
TRAIN_CLIP_THR = globals().get('TRAIN_CLIP_THR', CLIP_EVAL_THR)
if TRAIN_NORMAL_ONLY:
    if TARGET_MODE == 'log_diff' and 'revenue_change_next' in df_train.columns:
        y_train_eval = df_train['revenue_change_next']
    else:
        y_train_eval = y_train
    train_mask = y_train_eval.notna() & (y_train_eval.abs() <= TRAIN_CLIP_THR)
    # filter training matrices + targets
    if isinstance(X_tr_no_nan, pd.DataFrame):
        X_tr_no_nan = X_tr_no_nan.loc[train_mask]
    else:
        X_tr_no_nan = X_tr_no_nan[train_mask.values]

    if hasattr(Y_tr_t, 'loc'):
        Y_tr_t = Y_tr_t.loc[train_mask]
    else:
        Y_tr_t = Y_tr_t[train_mask.values]

    # keep y_train_clip aligned for weights
    y_train_clip = Y_tr_t

    # filter df_train for CV years alignment
    df_train_cv = df_train.loc[train_mask].copy()
else:
    df_train_cv = df_train.copy()

# Tail-aware sample weights (train only)
sample_weights = 1.0 / (np.abs(y_train_clip) + WEIGHT_EPS)
# clip weights to avoid extremes
sample_weights = np.clip(sample_weights, WEIGHT_CLIP_LO, WEIGHT_CLIP_HI)


# Custom MAPE scorer (exclude zeros)
def mape_thresholded(y_true, y_pred, thr=NEAR_ZERO_THR):
    y_true = np.asarray(y_true)
    y_pred = np.asarray(y_pred)
    mask = np.abs(y_true) >= thr
    if mask.sum() == 0:
        return np.nan, 0.0
    mape = np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask])) * 100
    coverage = mask.mean() * 100
    return mape, coverage

def mape_safe(y_true, y_pred):
    y_true = np.asarray(y_true)
    y_pred = np.asarray(y_pred)
    mask = y_true != 0
    if mask.sum() == 0:
        return np.nan
    return np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask])) * 100


# Clipped metrics on |y| <= CLIP_EVAL_THR (normal-band)
def clipped_metrics(y_true, y_pred, thr=CLIP_EVAL_THR):
    y_true = np.asarray(y_true)
    y_pred = np.asarray(y_pred)
    mask = np.isfinite(y_true) & np.isfinite(y_pred) & (np.abs(y_true) <= thr)
    if mask.sum() == 0:
        return {'RMSE_clip': np.nan, 'MAE_clip': np.nan, 'MAPE_clip': np.nan, 'Clip_cov(%)': 0.0}
    rmse = mean_squared_error(y_true[mask], y_pred[mask], squared=False)
    mae = mean_absolute_error(y_true[mask], y_pred[mask])
    mape = eps_mape(y_true[mask], y_pred[mask], eps=NEAR_ZERO_THR)
    return {'RMSE_clip': rmse, 'MAE_clip': mae, 'MAPE_clip': mape, 'Clip_cov(%)': mask.mean() * 100}

mape_scorer = make_scorer(mape_safe, greater_is_better=False)

# Thresholded MAPE scorer for near-zero stability
def mape_thr_only(y_true, y_pred):
    mape, _ = mape_thresholded(y_true, y_pred, thr=NEAR_ZERO_THR)
    return mape

mape_thr_scorer = make_scorer(mape_thr_only, greater_is_better=False)

# Select tuning scorer
scorer = mape_thr_scorer if TUNING_SCORER == 'mape_thr' else mape_scorer

def eps_mape(y_true, y_pred, eps=0.02):
    denom = np.maximum(np.abs(y_true), eps)
    return np.mean(np.abs(y_true - y_pred) / denom) * 100


# Time-aware CV split including 2019 and 2020 validation folds
if 'fiscal_year' in df_train.columns:
    years = df_train_cv['fiscal_year'].astype(int)

    # Build two folds: train<=2018 -> val=2019, train<=2019 -> val=2020
    fold1 = years.apply(lambda y: 0 if y <= 2018 else (1 if y == 2019 else -1))
    fold2 = years.apply(lambda y: 0 if y <= 2019 else (1 if y == 2020 else -1))

    # Keep only rows used in either fold
    use_mask = (fold1 >= 0) | (fold2 >= 0)
    use_mask_values = use_mask.values

    if hasattr(X_tr_no_nan, 'loc'):
        X_tr_no_nan = X_tr_no_nan.loc[use_mask]
    else:
        X_tr_no_nan = X_tr_no_nan[use_mask_values]

    if hasattr(Y_tr_t, 'loc'):
        Y_tr_t = Y_tr_t.loc[use_mask]
    else:
        Y_tr_t = Y_tr_t[use_mask_values]

    if 'sample_weights' in locals():
        if hasattr(sample_weights, 'loc'):
            sample_weights = sample_weights.loc[use_mask]
        else:
            sample_weights = sample_weights[use_mask_values]

    # Combine folds into a multi-fold PredefinedSplit
    fold1 = fold1.loc[use_mask].reset_index(drop=True)
    fold2 = fold2.loc[use_mask].reset_index(drop=True)

    # Encode fold1 as split 0, fold2 as split 1 (ignore rows with -1)
    test_fold = fold1.copy()
    test_fold[(fold1 == -1) & (fold2 >= 0)] = 1

    cv = PredefinedSplit(test_fold=test_fold)
else:
    cv = TimeSeriesSplit(n_splits=3)


# Random Forest tuning
rf_param_dist = {
    'n_estimators': [300, 600, 900],
    'max_depth': [None, 12, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [5, 10, 20],
    'max_features': ['sqrt', 'log2']
}

rf_search = RandomizedSearchCV(
    RandomForestRegressor(random_state=RANDOM_STATE, n_jobs=-1),
    rf_param_dist,
    n_iter=15,
    scoring=scorer,
    cv=cv,
    random_state=RANDOM_STATE,
    n_jobs=-1
)
rf_search.fit(X_tr_no_nan, Y_tr_t, sample_weight=sample_weights, groups=groups if "groups" in locals() else None)
rf_best = rf_search.best_estimator_

# Optional: XGBoost tuning (if available)
try:
    from xgboost import XGBRegressor
    xgb_param_dist = {
        'n_estimators': [400, 800, 1200],
        'learning_rate': [0.02, 0.05, 0.08],
        'max_depth': [2, 3, 4],
        'min_child_weight': [5, 10, 20],
        'subsample': [0.7, 0.85, 1.0],
        'colsample_bytree': [0.7, 0.85, 1.0],
        'reg_alpha': [0, 0.1, 1.0],
        'reg_lambda': [1.0, 5.0, 10.0]
    }
    xgb_search = RandomizedSearchCV(
        XGBRegressor(random_state=RANDOM_STATE, n_jobs=-1),
        xgb_param_dist,
        n_iter=12,
        scoring=scorer,
        cv=cv,
        random_state=RANDOM_STATE,
        n_jobs=-1
    )
    xgb_search.fit(X_tr_no_nan, Y_tr_t, sample_weight=sample_weights, groups=groups if "groups" in locals() else None)
    xgb_best = xgb_search.best_estimator_
except Exception as e:
    xgb_best = None
    print(f"XGBoost tuning skipped: {e}")

# Gradient Boosting tuning
gb_param_dist = {
    'n_estimators': [300, 600, 900],
    'learning_rate': [0.02, 0.05, 0.08],
    'max_depth': [2, 3, 4],
    'subsample': [0.7, 0.85, 1.0],
    'loss': ['absolute_error', 'huber']
}

gb_search = RandomizedSearchCV(
    GradientBoostingRegressor(random_state=RANDOM_STATE),
    gb_param_dist,
    n_iter=12,
    scoring=scorer,
    cv=cv,
    random_state=RANDOM_STATE,
    n_jobs=-1
)

gb_search.fit(X_tr_no_nan, Y_tr_t, sample_weight=sample_weights, groups=groups if "groups" in locals() else None)
gb_best = gb_search.best_estimator_


# HistGradientBoosting tuning (robust + strong on skewed targets)
hgb_param_dist = {
    'max_depth': [3, 5, 7],
    'learning_rate': [0.02, 0.05, 0.08],
    'max_iter': [300, 600, 900],
    'min_samples_leaf': [20, 50, 100],
    'l2_regularization': [0.1, 1.0, 5.0],
    'max_bins': [64, 128, 255],
    'loss': ['absolute_error', 'squared_error']
}

hgb_search = RandomizedSearchCV(
    HistGradientBoostingRegressor(random_state=RANDOM_STATE),
    hgb_param_dist,
    n_iter=12,
    scoring=scorer,
    cv=cv,
    random_state=RANDOM_STATE,
    n_jobs=-1
)

hgb_search.fit(X_tr_no_nan, Y_tr_t, sample_weight=sample_weights, groups=groups if 'groups' in locals() else None)
hgb_best = hgb_search.best_estimator_
# Metrics helper

def smape(y_true, y_pred):
    denom = (np.abs(y_true) + np.abs(y_pred)) / 2.0
    return np.mean(np.where(denom == 0, 0, np.abs(y_true - y_pred) / denom)) * 100


def directional_accuracy(y_true, y_pred):
    y_true = np.asarray(y_true)
    y_pred = np.asarray(y_pred)
    mask = np.isfinite(y_true) & np.isfinite(y_pred)
    if mask.sum() == 0:
        return np.nan
    return np.mean(np.sign(y_true[mask]) == np.sign(y_pred[mask])) * 100




def eval_metrics(model, X, y_true, transformed=True, calibrate=False):
    EVAL_CLIP_ONLY = globals().get('EVAL_CLIP_ONLY', False)
    preds_t = model.predict(X)
    preds_t = np.asarray(preds_t)
    if not np.isfinite(preds_t).all():
        med = np.nanmedian(preds_t)
        preds_t = np.nan_to_num(preds_t, nan=med, posinf=med, neginf=med)

    if TARGET_MODE == 'log_diff':
        preds = logdiff_to_revenue_change(pv_now_val, preds_t)
        y_eval = y_val_eval
    else:
        preds = inverse_target(preds_t, TARGET_TRANSFORM, yj) if transformed else preds_t
        y_eval = y_true

    # optional linear calibration on predictions
    if calibrate:
        from sklearn.linear_model import LinearRegression
        lr = LinearRegression()
        lr.fit(preds.reshape(-1, 1), y_eval)
        preds = lr.predict(preds.reshape(-1, 1))
    clip = clipped_metrics(y_eval, preds, thr=CLIP_EVAL_THR)
    if EVAL_CLIP_ONLY:
        mask = np.isfinite(y_eval) & np.isfinite(preds) & (np.abs(y_eval) <= CLIP_EVAL_THR)
        y_eval = y_eval[mask]
        preds = preds[mask]


    rmse = mean_squared_error(y_eval, preds, squared=False)
    mae = mean_absolute_error(y_eval, preds)
    mdae = median_absolute_error(y_eval, preds)
    mape = mape_safe(y_eval, preds)
    mape_thr, mape_cov = mape_thresholded(y_eval, preds, thr=NEAR_ZERO_THR)
    r2 = r2_score(y_eval, preds)

    # tail metrics
    tail_mask = np.abs(y_eval) > 50
    if tail_mask.any():
        tail_rmse = mean_squared_error(y_eval[tail_mask], preds[tail_mask], squared=False)
        tail_mae = mean_absolute_error(y_eval[tail_mask], preds[tail_mask])
    else:
        tail_rmse, tail_mae = np.nan, np.nan

    return rmse, mae, mdae, mape, mape_thr, mape_cov, r2, smape(y_eval, preds), directional_accuracy(y_eval, preds), tail_rmse, tail_mae, clip





results = []

if 'hgb_best' in locals():
    rmse, mae, mdae, mape, mape_thr, mape_cov, r2, sm, da, trmse, tmae, clip = eval_metrics(hgb_best, X_va_no_nan, y_val, transformed=False)
    results.append({'Model': 'HistGradientBoosting (tuned)', 'RMSE': rmse, 'MAE': mae, 'MdAE': mdae, 'MAPE': mape, 'MAPE_thr': mape_thr, 'MAPE_cov(%)': mape_cov, 'SMAPE': sm, 'R2': r2, 'DirectionalAcc(%)': da, 'Tail_RMSE': trmse, 'Tail_MAE': tmae, **clip})

rmse, mae, mdae, mape, mape_thr, mape_cov, r2, sm, da, trmse, tmae, clip = eval_metrics(rf_best, X_va_no_nan, y_val, transformed=False)
results.append({'Model': 'RandomForest (tuned)', 'RMSE': rmse, 'MAE': mae, 'MdAE': mdae, 'MAPE': mape, 'MAPE_thr': mape_thr, 'MAPE_cov(%)': mape_cov, 'SMAPE': sm, 'R2': r2, 'DirectionalAcc(%)': da, 'Tail_RMSE': trmse, 'Tail_MAE': tmae, **clip})

rmse, mae, mdae, mape, mape_thr, mape_cov, r2, sm, da, trmse, tmae, clip = eval_metrics(gb_best, X_va_no_nan, y_val, transformed=False)
results.append({'Model': 'GradientBoosting (tuned)', 'RMSE': rmse, 'MAE': mae, 'MdAE': mdae, 'MAPE': mape, 'MAPE_thr': mape_thr, 'MAPE_cov(%)': mape_cov, 'SMAPE': sm, 'R2': r2, 'DirectionalAcc(%)': da, 'Tail_RMSE': trmse, 'Tail_MAE': tmae, **clip})

if xgb_best is not None:
    rmse, mae, mdae, mape, mape_thr, mape_cov, r2, sm, da, trmse, tmae, clip = eval_metrics(xgb_best, X_va_no_nan, y_val, transformed=False)
    results.append({'Model': 'XGBoost (tuned)', 'RMSE': rmse, 'MAE': mae, 'MdAE': mdae, 'MAPE': mape, 'MAPE_thr': mape_thr, 'MAPE_cov(%)': mape_cov, 'SMAPE': sm, 'R2': r2, 'DirectionalAcc(%)': da, 'Tail_RMSE': trmse, 'Tail_MAE': tmae, **clip})

results_df = pd.DataFrame(results).sort_values('RMSE')
print("Tuned models evaluated on validation set (original scale, tail metrics included):")
display(results_df)

print("Hyperparameter tuning complete!")






# Feature importance (tuned models)
try:
    if isinstance(X_tr, pd.DataFrame):
        feature_names = X_tr.columns.tolist()
    else:
        feature_names = X_train.columns.tolist() if 'X_train' in globals() else [f'f{i}' for i in range(X_tr.shape[1])]

    def print_and_plot_importance(model, title, top_n=15):
        if not hasattr(model, 'feature_importances_'):
            print(f"{title}: no feature_importances_ attribute")
            return
        importances = model.feature_importances_
        fi = pd.DataFrame({'feature': feature_names, 'importance': importances})
        fi = fi.sort_values('importance', ascending=False).head(top_n)
        print(f"{title} (top {top_n}):")
        print(fi.to_string(index=False))

        plt.figure(figsize=(8, 6))
        sns.barplot(data=fi, x='importance', y='feature', color='steelblue')
        plt.title(title)
        plt.tight_layout()
        plt.show()

    print_and_plot_importance(rf_best, 'RandomForest (tuned) Feature Importance')
    print_and_plot_importance(gb_best, 'GradientBoosting (tuned) Feature Importance')
    if xgb_best is not None:
        print_and_plot_importance(xgb_best, 'XGBoost (tuned) Feature Importance')
except Exception as e:
    print(f"Feature importance skipped: {e}")
```

```python

# --- Final training + inference helpers ---
import pandas as pd
import numpy as np

USE_EXTERNAL_TEST = False  # set True for test_features.csv (2022-2023)
FINAL_MODEL = 'xgb'  # 'xgb' or 'lgb'

if USE_EXTERNAL_TEST:
    test_df = pd.read_csv('../data/processed/test_features.csv')
    full_plus_test = pd.concat([full_df, test_df], ignore_index=True, sort=False)
else:
    full_plus_test = full_df.copy()

# Build features (uses all years available for lag/target construction)
full_plus_test = build_panel_features(full_plus_test, verbose=False)


def add_final_features(train_df, test_df):
    """Add sector/company priors + normalization for final training."""
    train_df = train_df.copy()
    test_df = test_df.copy()

    # Sector-year median growth lag1 (train-only stats)
    if {'ateco_sector', 'fiscal_year', 'revenue_change_clean'} <= set(train_df.columns):
        sector_median = (
            train_df.groupby(['ateco_sector', 'fiscal_year'])['revenue_change_clean'].median()
        )

        def add_sector_median_lag1(df):
            df = df.copy()
            year_lag1 = df['fiscal_year'] - 1
            idx = list(zip(df['ateco_sector'], year_lag1))
            df['sector_median_growth_lag1'] = [sector_median.get(k, np.nan) for k in idx]
            if 'revchg_lag1' in df.columns:
                df['growth_excess'] = df['revchg_lag1'] - df['sector_median_growth_lag1']
            return df

        train_df = add_sector_median_lag1(train_df)
        test_df = add_sector_median_lag1(test_df)

    # Sector shock prior (train-only)
    if {'ateco_sector', 'revchg_lag1'} <= set(train_df.columns):
        sector_shock_prior = train_df.groupby('ateco_sector')['revchg_lag1'].median()
        train_df['sector_shock_prior'] = train_df['ateco_sector'].map(sector_shock_prior)
        test_df['sector_shock_prior'] = test_df['ateco_sector'].map(sector_shock_prior)
        train_df['company_vs_sector_shock'] = train_df['revchg_lag1'] - train_df['sector_shock_prior']
        test_df['company_vs_sector_shock'] = test_df['revchg_lag1'] - test_df['sector_shock_prior']

    # Company prior (train-only)
    if {'company_id', 'revenue_change_clean'} <= set(train_df.columns):
        company_prior = train_df.groupby('company_id')['revenue_change_clean'].median()
        train_df['company_prior'] = train_df['company_id'].map(company_prior)
        test_df['company_prior'] = test_df['company_id'].map(company_prior)
        if 'sector_median_growth_lag1' in train_df.columns:
            train_df['company_prior'] = train_df['company_prior'].fillna(train_df['sector_median_growth_lag1'])
            test_df['company_prior'] = test_df['company_prior'].fillna(test_df['sector_median_growth_lag1'])

    # Base-small flag (train-only threshold)
    if 'log_pv_lag1' in train_df.columns:
        thr = train_df['log_pv_lag1'].quantile(BASE_SMALL_Q)
        train_df['base_small_flag'] = (train_df['log_pv_lag1'] < thr).astype(int)
        test_df['base_small_flag'] = (test_df['log_pv_lag1'] < thr).astype(int)

    # Structural change z-score (train-only)
    if 'structural_change_raw' in train_df.columns:
        mu = train_df['structural_change_raw'].mean()
        sd = train_df['structural_change_raw'].std()
        if sd and sd > 0:
            train_df['structural_change_z'] = (train_df['structural_change_raw'] - mu) / sd
            test_df['structural_change_z'] = (test_df['structural_change_raw'] - mu) / sd
        else:
            train_df['structural_change_z'] = 0.0
            test_df['structural_change_z'] = 0.0

    return train_df, test_df


def train_and_predict(train_years, test_years, label):
    # Train rows with labels
    df_train = full_plus_test[
        full_plus_test['fiscal_year'].isin(train_years) & full_plus_test[TARGET_COL].notna()
    ].copy()

    # Test rows (features only)
    df_test = full_plus_test[full_plus_test['fiscal_year'].isin(test_years)].copy()
    df_test[TARGET_COL] = np.nan  # no peeking

    print(f"\n[{label}] Train years: {train_years}  Test years: {test_years} -> target years {[y + TARGET_SHIFT for y in test_years]}")

    df_train, df_test = add_final_features(df_train, df_test)

    # Build matrices (dummy val row to reuse pipeline)
    df_val_dummy = df_train.iloc[:1].copy()
    art = build_matrices(df_train, df_val_dummy, df_test_in=df_test)
    X_train_final = art['X_train']
    y_train_final = art['y_train']
    X_test_final = art['X_test']

    # Train model
    model = None
    if FINAL_MODEL == 'lgb':
        try:
            from lightgbm import LGBMRegressor
            model = LGBMRegressor(
                n_estimators=1200,
                learning_rate=0.03,
                num_leaves=31,
                max_depth=-1,
                subsample=0.8,
                colsample_bytree=0.8,
                reg_lambda=1.0,
                objective='regression_l1',
                random_state=RANDOM_STATE
            )
        except Exception as e:
            print(f"LightGBM failed ({e}); falling back to XGBoost")
            model = None

    if model is None:
        from xgboost import XGBRegressor
        model = XGBRegressor(
            n_estimators=900,
            learning_rate=0.05,
            max_depth=4,
            subsample=0.8,
            colsample_bytree=0.8,
            objective='reg:absoluteerror',
            random_state=RANDOM_STATE,
            n_jobs=-1
        )

    model.fit(X_train_final, y_train_final)
    preds = model.predict(X_test_final)

    pred_df = df_test[['company_id', 'fiscal_year']].copy()
    pred_df['revenue_change_next_year_pred'] = preds
    print('Prediction sample:')
    display(pred_df.head())
    return pred_df


# --- Scenario A: 2021 -> 2022 (holdout) ---
# Uses 2018-2020 for training; predicts 2021
pred_2021 = train_and_predict(train_years=[2018, 2019, 2020], test_years=[2021], label='2021->2022')

# --- Scenario B: 2022 -> 2023 (requires external test features) ---
if USE_EXTERNAL_TEST:
    # include 2021 in training (labels from 2022 production_value are available in test_features)
    pred_2022 = train_and_predict(train_years=[2018, 2019, 2020, 2021], test_years=[2022], label='2022->2023')


# --- Optional 2022->2023 metric evaluation using 2023 production_value ---
if USE_EXTERNAL_TEST:
    try:
        # Use test_features (2022-2023) as evaluation source
        test_df_eval = pd.read_csv('../data/processed/test_features.csv')

        pv_2022 = test_df_eval[test_df_eval['fiscal_year'] == 2022][['company_id', 'production_value']].copy()
        pv_2023 = test_df_eval[test_df_eval['fiscal_year'] == 2023][['company_id', 'production_value']].copy()

        eval_df = pred_2022.merge(pv_2022, on='company_id', how='left')
        eval_df = eval_df.merge(pv_2023, on='company_id', how='left', suffixes=('_2022', '_2023'))

        pv_now = eval_df['production_value_2022']
        pv_next = eval_df['production_value_2023']
        eval_df['y_true_2022'] = np.where(pv_now > 0, (pv_next - pv_now) / pv_now * 100.0, np.nan)

        mask = eval_df['y_true_2022'].notna() & np.isfinite(eval_df['y_true_2022'])
        y_true = eval_df.loc[mask, 'y_true_2022']
        y_pred = eval_df.loc[mask, 'revenue_change_next_year_pred']

        if len(y_true) == 0:
            print('No overlapping companies with 2023 production_value. Metrics not available.')
        else:
            from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
            rmse = mean_squared_error(y_true, y_pred, squared=False)
            mae = mean_absolute_error(y_true, y_pred)
            r2 = r2_score(y_true, y_pred)
            mape = np.mean(np.abs((y_true - y_pred) / y_true).replace([np.inf, -np.inf], np.nan).dropna()) * 100

            print('2022 Holdout Metrics (using 2023 production_value for ground truth):')
            print(f'RMSE: {rmse:.4f}  MAE: {mae:.4f}  MAPE: {mape:.4f}  R2: {r2:.4f}')
    except Exception as e:
        print(f'2022->2023 evaluation failed: {e}')
```
